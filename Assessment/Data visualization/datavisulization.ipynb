#### Q.1 Convert a 1D array to a 2D array with 2 rows

```python
import numpy as np

a = np.array([1, 2, 3, 4, 5, 6])

result = a.reshape(2, 3)

print(result)
```

#### Q.2 Get the common items between a and b

```python
import numpy as np

a = np.array([1,2,3,2,3,4,3,4,5,6])
b = np.array([7,2,10,2,7,4,9,4,9,8])

result = np.intersect1d(a, b)

print(result)
```

#### Q.3 Get all items between 5 and 10 from a

```python
import numpy as np

a = np.array([2, 6, 1, 9, 10, 3, 27])

result = a[(a >= 5) & (a <= 10)]

print(result)
```

#### Q.4 Limit the number of items printed in NumPy array

```python
import numpy as np

np.set_printoptions(threshold=6)

a = np.arange(15)

print(a)
```

#### 1. Compute minimum, 25th percentile, median, 75th percentile and maximum

```python
import pandas as pd
import numpy as np

ser = pd.Series(np.random.normal(10, 5, 25))

print("Minimum :", ser.min())
print("25th Percentile :", ser.quantile(0.25))
print("Median :", ser.median())
print("75th Percentile :", ser.quantile(0.75))
print("Maximum :", ser.max())
```

### MATPLOTLIB


#### 9. Create line plot


```python
import matplotlib.pyplot as plt

x = [3, 4, 5, 6]
y = [1.5, 2, 2.5, 3]

plt.plot(x, y)

plt.xlabel("X")
plt.ylabel("Y")

plt.title("Line Plot")

plt.show()
```


#### 10. Plot x, x², x³ and x⁴

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.3)

plt.plot(x, x**1, 'ro')
plt.plot(x, x**2, 'bs')
plt.plot(x, x**3, 'g')
plt.plot(x, x**4, ':')

plt.xlim(0, 6)
plt.ylim(0, 125)


plt.show()
```


#### 11. Bar plot for height comparison

```python
import matplotlib.pyplot as plt

height = [179, 155, 191, 152, 188, 177]
names = ['QA', 'WB', 'EC', 'RD', 'TE', 'YF']

plt.bar(names, height)

plt.title("Height Comparison")

plt.xlabel("Names")
plt.ylabel("Height")

plt.show()
```

### MACHINE LEARNING

#### 15. Supervised vs Unsupervised Machine Learning

##### Supervised Learning

##### Supervised learning is a type of machine learning in which the model is trained using labeled data.

- Labeled data means the input data already has the correct output.

- The model learns from the training data and predicts outputs for new data.

#### Features
- Uses labeled data
- Output is already known
- Used for prediction tasks
- Higher accuracy in many cases

#### Types
- Classification
- Regression

#### Common Algorithms
- Linear Regression
- Logistic Regression
- Decision Tree
- Random Forest
- Support Vector Machine (SVM)

#### Real-Life Examples
- Email spam detection
- House price prediction
- Disease prediction
- Student result prediction


##### Unsupervised Learning

##### Unsupervised learning is a type of machine learning in which the model is trained using unlabeled data.

- The machine tries to find hidden patterns, relationships, or groups in the data by itself.

#### Features
- Uses unlabeled data
- Output is not known
- Used for pattern detection
- Helps in data grouping

#### Types
- Clustering
- Association

#### Common Algorithms
- K-Means Clustering
- DBSCAN
- Hierarchical Clustering
- Apriori Algorithm

#### Real-Life Examples
- Customer segmentation
- Product recommendation
- Fraud detection
- Market basket analysis
