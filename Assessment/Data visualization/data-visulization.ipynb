```python
#4. What’s the average age of patients (in years)?

import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
import numpy as np

sns.barplot(x = "cardio" , y="age" , data=health , estimator=np.mean)
plt.xlabel("cardio (0 = no , 1 = yes)")
plt.ylabel("avg (years)")
plt.title ("average year by cardio")
plt.show()
```

```python
# 5 .What is the distribution of the target variable (cardio)?

plt.figure(figsize=(6,4))
sns.countplot(x="cardio", data=health)
plt.xlabel("Cardio (Target Variable)")
plt.ylabel("Count")
plt.title("Distribution of Target Variable (cardio)")
plt.show()
```

```python
# 6 .What is the average age (in years) of patients?

sns.barplot(x = "cardio" , y = "age" ,  data=health , estimator=np.mean)
plt.xlabel("cardio (0 = no , 1 = yes)")
plt.ylabel("age (avg age)")
plt.title ("average year by cardio")
plt.show()
```

```python
# 7. What is the distribution of BMI? (Create new BMI feature)

health["height_m"] = health["height"] / 100
health["BMI"] = health["weight"] / (health["height_m"] ** 2)

plt.Figure()
sns.histplot(health["BMI"] , bins=30)
plt.xlabel("BMI")
plt.ylabel("frequency")
plt.title("BMI Distribution")
plt.show()
```

```python
# 8.Are there outliers in height or weight?

plt.subplot(1,2,1)
sns.boxplot(y=health['height'], color="lightblue")
plt.title("Height")

plt.subplot(1,2,2)
sns.boxplot(y=health['weight'], color="lightgreen")
plt.title("Weight")
plt.show()
```
