```python
import bz2
import pandas as pd
import re

data = []
```

```python
with bz2.open('test.ft.txt.bz2', 'rt', encoding='utf-8') as file:

    for i, line in enumerate(file):

        if i == 5000:
            break

        parts = line.split(' ', 1)

        label = parts[0]
        review = parts[1] if len(parts) > 1 else ""

        data.append([label, review])

    df = pd.DataFrame(data, columns=['label', 'review'])

    print("Total Reviews:", len(df))

    print("\nMissing Values:")
    print(df.isnull().sum())

    print("\nDuplicate Reviews:")
    print(df.duplicated().sum())

    df['word_count'] = df['review'].apply(lambda x: len(str(x).split()))


    print("\nAverage Review Length:")
    print(df['word_count'].mean())
    
    print("\nCategory Count:")
    print(df['label'].value_counts())
```
