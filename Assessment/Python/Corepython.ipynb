### 1. Fruit Manager

```python

stock = {}


def add_fruit():

    fruit_name = input("Enter Fruit Name : ").title()

    try:
        quantity = int(input("Enter Quantity : "))
        price = int(input("Enter Price : "))

        stock[fruit_name] = {
            "quantity": quantity,
            "price": price
        }

        print("\nFruit Added Successfully")

    except ValueError:
        print("\nPlease Enter Valid Number")



def view_stock():

    if not stock:
        print("\nNo Stock Available")

    else:

        print("\n======= AVAILABLE STOCK =======")

        for fruit, details in stock.items():

            print(
                f"{fruit} -> Quantity : {details['quantity']} | Price : {details['price']}"
            )


def update_stock():

    fruit_name = input("Enter Fruit Name : ").title()

    if fruit_name in stock:

        try:
            new_quantity = int(input("Enter New Quantity : "))
            new_price = int(input("Enter New Price : "))

            stock[fruit_name]["quantity"] = new_quantity
            stock[fruit_name]["price"] = new_price

            print("\nStock Updated Successfully")

        except ValueError:
            print("\nInvalid Input")

    else:
        print("\nFruit Not Found")



def buy_fruit():

    fruit_name = input("Enter Fruit Name : ").title()

    if fruit_name in stock:

        try:
            buy_quantity = int(input("Enter Quantity : "))

            available_quantity = stock[fruit_name]["quantity"]
            fruit_price = stock[fruit_name]["price"]

            if buy_quantity <= available_quantity:

                total_bill = buy_quantity * fruit_price

                stock[fruit_name]["quantity"] -= buy_quantity

                print("\n======= CUSTOMER BILL =======")

                print("Fruit Name :", fruit_name)
                print("Fruit Price :", fruit_price)
                print("Quantity :", buy_quantity)
                print("Total Bill :", total_bill)

                # Store Transaction in File
                file = open("transactions.txt", "a")

                file.write(
                    f"Fruit : {fruit_name}, "
                    f"Quantity : {buy_quantity}, "
                    f"Total Bill : {total_bill}\n"
                )

                file.close()

                print("\nPurchase Successful")

            else:
                print("\nInsufficient Stock")

        except ValueError:
            print("\nPlease Enter Valid Quantity")

    else:
        print("\nFruit Not Available")



while True:

    print("\n==============================")
    print("      FRUIT STORE SYSTEM")
    print("==============================")

    print("\n1. Fruit Manager")
    print("2. Customer")
    print("3. Exit")

    choice = input("\nEnter Your Choice : ")

   

    if choice == "1":

        while True:

            print("\n======= MANAGER MENU =======")

            print("\n1. Add Fruit Stock")
            print("2. View Fruit Stock")
            print("3. Update Fruit Stock")
            print("4. Back")

            manager_choice = input("\nEnter Choice : ")

            if manager_choice == "1":

                add_fruit()

            elif manager_choice == "2":

                view_stock()

            elif manager_choice == "3":

                update_stock()

            elif manager_choice == "4":

                break

            else:
                print("\nInvalid Choice")



    elif choice == "2":

        while True:

            print("\n======= CUSTOMER MENU =======")

            print("\n1. Buy Fruit")
            print("2. View Fruit Stock")
            print("3. Back")

            customer_choice = input("\nEnter Choice : ")

            if customer_choice == "1":

                buy_fruit()

            elif customer_choice == "2":

                view_stock()

            elif customer_choice == "3":

                break

            else:
                print("\nInvalid Choice")


    elif choice == "3":

        print("\nThank You For Using Fruit Store System")
        break

    else:
        print("\nInvalid Choice")
```
