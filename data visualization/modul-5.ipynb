### 1. Write an essay: Importance of Data Cleaning in Data Science

Data is the foundation of data science. Every analysis, prediction, or decision made by data scientists depends on the quality of the data they use . This is where data cleaning plays a crucial role .Data cleaning, also known as data cleansing or data preprocessing,

Data cleaning is the process of preparing raw data for analysis by handling missing values, removing duplicates, correcting inconsistencies, and standardizing formats. It ensures that the dataset is accurate, consistent, and usable.

For example, if a dataset contains customer details, data cleaning may involve fixing spelling errors in names, filling missing age values, or converting date formats into a common standard.


### Why is Data Cleaning Important?

1. Improves Data Quality :- Data cleaning makes data more accurate and reliable by removing incorrect and inconsistent information.

2. Enhances Accuracy of  :- Clean data ensures models learn correct patterns for better predictions.

3. Saves Time and Resources :- Reduces errors and rework, speeding up analysis and decisions.

4. Supports Better Decision-Making :- Reliable data leads to valid and effective business decisions.

5. Improves Data Consistency :-Standardizes data from multiple sources for easier integration.

### Common Steps in Data Cleaning

1. Handling Missing Data – Filling missing values using averages, medians, or predictive methods.

2. Removing Duplicates – Deleting repeated records to avoid bias.

3. Correcting Errors – Fixing spelling mistakes, incorrect entries, or formatting issues.

4. Standardizing Data Formats – Ensuring consistent date, time, and unit formats.

5. Outlier Detection – Identifying and managing unusual data points that may distort analysis.

6. Data Validation – Checking that data meets the required rules and constraints.


Imagine a hospital maintaining patient records. If names are misspelled, dates of birth are missing, or addresses are inconsistent, it becomes difficult to analyze patient history or predict health trends. After data cleaning, the hospital can use the refined data for accurate diagnosis and efficient service delivery.

```python
import pandas  as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


cars = pd.read_csv('data.csv.zip')
```

```python
cars
```

```python
import seaborn as sns

cars . isnull() . sum()
sns . heatmap(cars . isnull())
```

```python
# Fill missing Engine HP with mean
cars.fillna(({'Engine HP':cars['Engine HP'].mean()}) , inplace=True)
```

```python
# Fill missing Engine Cylinders with median
cars.fillna(({'Engine Cylinders' : cars['Engine Cylinders'] . median()}) , inplace = True)
```

```python
# Fill missing Number of Doors with mode
cars.fillna(({'	Number of Doors' : cars['Number of Doors'] . mode()}) , inplace= True)
```

```python
cars.fillna(({'Market Category' : 'unknown'}), inplace = True)
```

```python
# Fill Engine Fuel Type with the most common value
cars.fillna(({'Engine Fuel Type' : cars['Engine Fuel Type'] . mode()[0]}) , inplace = True)
```

## Top 5 car makers

```python
import matplotlib.pyplot as plt

top_makes = cars['Make'].value_counts().head(5)
sns.barplot(x=top_makes.values, y=top_makes.index , color= "black")
```

### Distribution of Transmission Type Horsepower

```python

plt.figure(figsize=(8,5))
sns.histplot(cars['Engine HP'], bins=50, color='pink')
plt.title('Distribution of Transmission Type Horsepower')
plt.xlabel('Engine HP')
plt.ylabel('Count')
plt.show()
```

### Transmission Type Distribution 

```python

plt.figure(figsize=(10,5))
sns.countplot(x='Transmission Type', data=cars, palette='viridis')
plt.title('Transmission Type Distribution')
plt.show()
```

```python
# It is shown that front wheel drive is the most occured type.

plt.figure(figsize=(10,6))
sns.heatmap(cars.select_dtypes(include='number').corr(), annot=True, cmap='summer')
plt.title('Correlation Heatmap')
plt.show()
```

### This shows the correlation between numerical columns

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Vehicle Size', y='MSRP', data=cars, estimator='mean', ci=None , color= "lightgreen")
plt.show()
```

### the MSRP is getting high with the vehicle size.

```python

plt.figure(figsize=(8 , 5))
sns.boxenplot(x='Vehicle Size', y='Engine HP', data=cars , color= "red")
plt.title('Engine HP Distribution by Vehicle Size')
plt.show()
```

### We can see many outliers in the graph.

```python
yearly = cars.groupby('Year')['highway MPG'].mean().reset_index()
sns.lineplot(x='Year', y='highway MPG', data=yearly )
plt.title('Trend of Avg Highway MPG Over Years')
plt.show()
```

### It is seen that over the years, highway MPG is increasing.
