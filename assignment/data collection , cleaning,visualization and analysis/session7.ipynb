
1. Given a CSV file containing Zomato restaurant ratings, use pandas to detect outliers in the 'user_rating' column using the IQR method and print the indices of the detected outliers.

```python
import pandas as pd

df = pd.read_csv("zomato_ratings.csv")

Q1 = df["user_rating"].quantile(0.25)
Q3 = df["user_rating"].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

outliers = df[
    (df["user_rating"] < lower_limit) |
    (df["user_rating"] > upper_limit)
]

print("Outlier Indices:")
print(outliers.index.tolist())

print("\nOutlier Records:")
print(outliers)
```

3. Apply winsorization to the 'transaction_amount' column in a Paytm transactions DataFrame to cap all values above the 95th percentile and below the 5th percentile, then display the updated column statistics.

```python
import pandas as pd
from scipy.stats.mstats import winsorize

paytm_df = pd.DataFrame({
    "transaction_id": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "transaction_amount": [120, 250, 300, 450, 600, 750, 900, 1500, 5000, 10000]
})

print("Original DataFrame:")
print(paytm_df)

paytm_df["transaction_amount"] = winsorize(
    paytm_df["transaction_amount"],
    limits=(0.05, 0.05)
)

print("\nUpdated DataFrame After Winsorization:")
print(paytm_df)

# Display updated statistics
print("\nUpdated Column Statistics:")
print(paytm_df["transaction_amount"].describe())
```

4. You have a DataFrame of Flipkart product prices stored as strings with currency symbols (e.g., '₹1,299'). Convert this column to numeric type using pandas, ensuring all non-numeric characters are removed.<br><br><em><strong>Hint:</strong> Use str.replace() and astype().</em>

```python
import pandas as pd

# Create a sample Flipkart products DataFrame
flipkart_df = pd.DataFrame({
    "product_name": [
        "Smartphone",
        "Laptop",
        "Headphones",
        "Smart TV",
        "Washing Machine"
    ],
    "price": [
        "₹12,999",
        "₹54,999",
        "₹2,499",
        "₹39,999",
        "₹24,500"
    ]
})

print("Original DataFrame:")
print(flipkart_df)

# Remove ₹ symbol and commas, then convert to integer
flipkart_df["price"] = (
    flipkart_df["price"]
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(int)
)

print("\nUpdated DataFrame:")
print(flipkart_df)

print("\nData Types:")
print(flipkart_df.dtypes)
```

5. Fix the following code snippet where the 'is_premium' column in a Spotify user DataFrame is a mix of boolean, string, and integer types. Convert the entire column to boolean type, treating 'True', 1, and 'yes' as True, and everything else as False.

```python
import pandas as pd

# Create a sample Spotify users DataFrame
spotify_df = pd.DataFrame({
    "user_name": ["Amit", "Priya", "Rahul", "Neha", "Karan", "Riya", "Jay"],
    "is_premium": [True, "True", 1, "yes", False, 0, "No"]
})

print("Original DataFrame:")
print(spotify_df)

# Convert all values to boolean
spotify_df["is_premium"] = (
    spotify_df["is_premium"]
    .astype(str)                 # Convert everything to string
    .str.lower()                 # Convert to lowercase
    .isin(["true", "1", "yes"])  # Treat these values as True
)

print("\nUpdated DataFrame:")
print(spotify_df)

print("\nData Types:")
print(spotify_df.dtypes)
```
