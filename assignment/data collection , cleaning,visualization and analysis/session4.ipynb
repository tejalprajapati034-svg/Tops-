#### 4.Download a large Excel file (at least 10,000 rows) of Flipkart product listings. Use pd.read_excel() with the chunksize parameter to read the file in chunks of 2000 rows, and print the number of rows in each chunk as you iterate.

```python
import pandas as pd

data = {
    "Product_ID": [1001, 1002, 1003, 1004, 1005],
    "Product_Name": ["iPhone 15", "Galaxy S24", "OnePlus 12", "HP Laptop", "Dell Inspiron"],
    "Category": ["Mobile", "Mobile", "Mobile", "Laptop", "Laptop"],
    "Brand": ["Apple", "Samsung", "OnePlus", "HP", "Dell"],
    "Price": [79999, 74999, 64999, 55999, 52999],
    "Rating": [4.8, 4.7, 4.6, 4.5, 4.4],
    "Stock": [25, 30, 20, 15, 18]
}

df = pd.DataFrame(data)

df.to_excel("flipkart_products.xlsx", index=False)


chunk_size = 2000

for start in range(0, len(df), chunk_size):
    chunk = df.iloc[start:start + chunk_size]
    print("Rows in chunk:", len(chunk))
```

#### 5.Use ChatGPT or Copilot to generate a Python code snippet that reads a semicolon-separated CSV of Paytm transactions, detects missing values, and prints out which columns have nulls. Test the code with a small sample file and fix any errors you encounter.

```python
import pandas as pd

df = pd.read_csv("paytm_transactions.csv", sep=";")

print("Paytm Transactions:")
print(df)

print("\nMissing Values in Each Column:")
print(df.isnull().sum())

print("\nColumns with Null Values:")
null_columns = df.columns[df.isnull().any()]
print(list(null_columns))
```
