
1. Write a short Python script that generates a random dataset of 10 users, each with two features: number of daily app opens and average session duration (in minutes). Print the dataset as a list of dictionaries.

```python
import random

users = []

for i in range(1, 11):
    user = {
        "User_ID": f"U{i}",
        "Daily_App_Opens": random.randint(5, 30),
        "Average_Session_Duration": random.randint(5, 60)
    }
    users.append(user)

print(users)
```

2. Create a table comparing supervised and unsupervised learning in at least 4 points, using real-world app examples (like Instagram feed ranking for supervised, Spotify playlist grouping for unsupervised).

####  Supervised Learning
- Uses labeled data (input with known output).
- Predict or classify future outcomes.
- Instagram ranks your feed based on your previous likes, comments, and shares to predict which posts you'll engage with.
- Produces a predicted value or category (e.g., spam/not spam, price prediction).
- Linear Regression, Logistic Regression, Decision Tree, Random Forest, K-Nearest Neighbors (KNN).

#### Unsupervised Learning
- Uses unlabeled data (only input data, no known output).
- Discover hidden patterns, relationships, or groups in data.
- Spotify automatically groups similar songs into playlists like "Chill", "Workout", or "Party" based on audio features.
- Produces clusters or groups of similar data points without predefined labels.
- K-Means Clustering, Hierarchical Clustering, DBSCAN, PCA.
