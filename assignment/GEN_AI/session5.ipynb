2.Create a simple Q&A chatbot for Zomato-style restaurant info using any language or platform you know (Python, JavaScript, etc.): given a restaurant name and cuisine, return a canned response like 'Yes, Pizza Palace serves Italian cuisine.'

```python
# Simple Zomato-style Restaurant Q&A Chatbot

restaurants = {
    "Pizza Palace": "Italian",
    "Spice Garden": "Indian",
    "Dragon House": "Chinese",
    "Burger Point": "Fast Food"
}

restaurant_name = input("Enter restaurant name: ")
cuisine = input("Enter cuisine: ")

if restaurant_name in restaurants:
    if restaurants[restaurant_name].lower() == cuisine.lower():
        print(f"Yes, {restaurant_name} serves {cuisine} cuisine.")
    else:
        print(f"No, {restaurant_name} does not serve {cuisine} cuisine.")
else:
    print("Restaurant not found in database.")
```

Use GitHub Copilot or ChatGPT to help you write a function that takes a list of song names and returns only those that start with the letter 'S', like a Spotify playlist filter.<br><br><em><strong>Constraint:</strong> You must show the prompt you used and the code Copilot or ChatGPT generated.</em>

promt : Write a Python function that takes a list of song names and returns only the songs that start with the letter 'S', similar to a Spotify playlist filter.

```python
def filter_songs(song_list):
    return [song for song in song_list if song.startswith('S')]

# Example playlist
songs = [
    "Shape of You",
    "Believer",
    "Sunflower",
    "Senorita",
    "Perfect",
    "Stay"
]

result = filter_songs(songs)

print("Songs starting with 'S':")
print(result)
```
