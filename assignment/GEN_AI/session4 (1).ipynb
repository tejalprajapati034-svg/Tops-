1.
Simulate the forward diffusion process by writing a Python function that takes a grayscale image (use any small PNG or JPEG) and adds Gaussian noise to it for 10 steps, saving each noisy image to disk.

```python
import cv2
import numpy as np
```

```python
image = cv2.imread("tree.png", cv2.IMREAD_GRAYSCALE)
```

```python
if image is None:
    raise FileNotFoundError(
        "Error: tree.png not found."
    )
```

```python
current_image = image.copy()
```

```python
for step in range(1, 11):

    noise = np.random.normal(
        0,
        15,
        current_image.shape
    )

    noisy_image = current_image + noise

    noisy_image = np.clip(
        noisy_image,
        0,
        255
    ).astype(np.uint8)

    cv2.imwrite(
        f"noisy_step_{step}.png",
        noisy_image
    )

    current_image = noisy_image

print("10 noisy images saved successfully.")

```

2.
Write a Python script that takes a noisy image and applies a simple denoising filter (like median blur or Gaussian blur) to approximate the reverse diffusion process, then compare the denoised output visually with the original image.<br><br><em><strong>Hint:</strong> Use OpenCV's cv2.medianBlur() or cv2.GaussianBlur() for denoising.</em>

```python
import cv2
import matplotlib.pyplot as plt
```

```python
original = cv2.imread("tree.png", cv2.IMREAD_GRAYSCALE)
```

```python
noisy = cv2.imread("noisy_step_10.png", cv2.IMREAD_GRAYSCALE)
```

```python
if original is None or noisy is None:
    raise FileNotFoundError("Image not found.")
```

```python
denoised = cv2.GaussianBlur(noisy, (5, 5), 0)
```

```python
cv2.imwrite("denoised_image.png", denoised)

print("Denoised image saved successfully!")
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(original, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(noisy, cmap='gray')
plt.title("Noisy Image")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(denoised, cmap='gray')
plt.title("Denoised Image")
plt.axis("off")

plt.show()
```
