Implement a simple RNN cell from scratch in Python that takes a sequence of 5 numbers (e.g., song play counts per day from Spotify) and outputs the final hidden state after processing the sequence.<br><br><em><strong>Hint:</strong> Use the formula: h_t = tanh(Wx_t + Uh_{t-1} + b), where h_0 = 0.</em>

```python
import numpy as np

# Sequence (song play counts)
sequence = [12, 15, 10, 18, 20]

# Initialize weights and bias
W = 0.5
U = 0.8
b = 0.1

# Initial hidden state
h = 0

# Process sequence
for x in sequence:
    h = np.tanh(W * x + U * h + b)
    print("Hidden State:", h)

print("\nFinal Hidden State:", h)
```

3. Given the following sentence: 'I love cricket and IPL matches', convert each word to a unique integer (word-to-index mapping), then simulate how an RNN would process this sequence by updating and printing the hidden state at each step.

```python
import numpy as np

sentence = "I love cricket and IPL matches"

# Split sentence
words = sentence.split()

# Word-to-index mapping
word_to_index = {
    word: i + 1
    for i, word in enumerate(words)
}

print("Word to Index Mapping:")
print(word_to_index)

# Convert sentence to sequence
sequence = [word_to_index[word] for word in words]

# Simple RNN parameters
W = 0.4
U = 0.6
b = 0.1

h = 0

print("\nHidden States:")

for x in sequence:
    h = np.tanh(W * x + U * h + b)
    print(h)
```

4. Explain the vanishing gradient problem in RNNs using your own words, and give one real-world example (from text, speech, or time-series apps) where this issue might cause poor predictions.

#### Answer:

- The vanishing gradient problem occurs when gradients become extremely small during training, making it difficult for the RNN to learn long-term dependencies. As a result, the model gradually forgets information from earlier time steps.

#### Real-World Example:

- In a speech recognition application, an RNN may forget the beginning of a long spoken sentence before reaching the end. This can lead to incorrect transcription because earlier words are no longer influencing the prediction.

5. Use ChatGPT to generate a short Python code snippet that demonstrates Backpropagation Through Time (BPTT) for a 3-step RNN sequence. Paste the code and briefly describe what each part does.

```python
import numpy as np

# Input sequence
x = [1, 2, 3]

# Parameters
W = 0.5
U = 0.8
b = 0.1

# Initial hidden state
h = 0
hidden_states = []

# Forward Pass
for t in x:
    h = np.tanh(W * t + U * h + b)
    hidden_states.append(h)

print("Hidden States:", hidden_states)

# Dummy gradient from output
grad = 1

print("\nBackpropagation Through Time")

# Backward Pass
for h in reversed(hidden_states):
    grad = grad * (1 - h**2) * U
    print("Gradient:", grad)
```
