1. Download a small set of 20-30 fashion item images (t-shirts, shoes, bags) from the web and organize them into separate folders by category (e.g., t_shirt, shoe, bag) to use as your dataset for CNN training.

- To prepare a dataset for training a Convolutional Neural Network (CNN), I collected 20–30 fashion item images from free online sources such as Pexels, Pixabay, or Unsplash. The images were divided into three categories:

- T-Shirt
- Shoe
- Bag

#### A basic CNN model consists of the following layers:

- Conv2D Layer: Extracts important features such as edges, textures, and shapes from the input images.
- MaxPooling2D Layer: Reduces the size of the feature maps, decreasing computation while keeping the most important features.
- Flatten Layer: Converts the 2D feature maps into a 1D vector.
- Dense Layer: Learns patterns from the extracted features and performs classification.
- Output Layer: Uses the Softmax activation function to predict the probability of each fashion category (T-shirt, Shoe, or Bag).

2. Build a basic Convolutional Neural Network using Keras with a Conv2D -> MaxPooling2D -> Flatten -> Dense stack to classify your fashion images into their respective categories.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Image Generator
train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

train_generator = train_datagen.flow_from_directory(
    "Fashion_Dataset/train",
    target_size=(128,128),
    batch_size=4,
    class_mode="categorical",
    subset="training"
)

validation_generator = train_datagen.flow_from_directory(
    "style/train",
    target_size=(128,128),
    batch_size=4,
    class_mode="categorical",
    subset="validation"
)

# CNN Model
model = Sequential()

model.add(
    Conv2D(
        32,
        (3,3),
        activation="relu",
        input_shape=(128,128,3)
    )
)

model.add(MaxPooling2D((2,2)))

model.add(Flatten())

model.add(Dense(64,activation="relu"))

model.add(Dense(3,activation="softmax"))

model.summary()
```

3. Train your baseline CNN model for 5 epochs and print the training and validation accuracy after each epoch.

```python
# Compile Model
model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

# Train Model
history = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=5
)

# Print Accuracy
print("\nTraining Accuracy:")
print(history.history["accuracy"])

print("\nValidation Accuracy:")
print(history.history["val_accuracy"])
```

4. Evaluate your trained model's accuracy on a separate test set of 6-10 unseen fashion images and display which images were misclassified.<br><br><em><strong>Hint:</strong> Use model.predict() and compare the predicted label with the actual folder/category.</em>

```python
import numpy as np
import os
from tensorflow.keras.preprocessing import image

# Class Labels
class_names = ["bag","shoe","t_shirt"]

test_folder = "Fashion_Dataset/test"

correct = 0
total = 0

print("\nMisclassified Images:\n")

for actual_class in class_names:

    folder = os.path.join(test_folder, actual_class)

    for file in os.listdir(folder):

        img_path = os.path.join(folder,file)

        img = image.load_img(
            img_path,
            target_size=(128,128)
        )

        img = image.img_to_array(img)/255.0

        img = np.expand_dims(img,axis=0)

        prediction = model.predict(img,verbose=0)

        predicted_class = class_names[np.argmax(prediction)]

        total += 1

        if predicted_class == actual_class:

            correct += 1

        else:

            print(
                file,
                " Actual:",
                actual_class,
                " Predicted:",
                predicted_class
            )

accuracy = correct/total*100

print("\nTest Accuracy:",accuracy,"%")
```
