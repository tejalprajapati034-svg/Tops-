1. Create a Python script using NumPy to implement a perceptron that learns to classify whether a Flipkart product review is 'positive' (1) or 'negative' (0) based on two features: review length and number of exclamation marks. Use a small manually created dataset for training.

```python
import numpy as np

# Features:
# [Review Length, Number of Exclamation Marks]

X = np.array([
    [120, 5],
    [100, 4],
    [20, 0],
    [35, 1],
    [150, 6],
    [40, 0]
])

# Labels
# 1 = Positive
# 0 = Negative
y = np.array([1, 1, 0, 0, 1, 0])

# Initialize weights and bias
weights = np.zeros(2)
bias = 0
learning_rate = 0.001

# Training
for epoch in range(100):
    for i in range(len(X)):
        weighted_sum = np.dot(X[i], weights) + bias
        prediction = 1 if weighted_sum >= 0 else 0

        error = y[i] - prediction

        weights += learning_rate * error * X[i]
        bias += learning_rate * error

print("Weights:", weights)
print("Bias:", bias)

print("\nPredictions:")

for sample in X:
    prediction = 1 if np.dot(sample, weights) + bias >= 0 else 0
    print(sample, "->", prediction)
```

2. Write a function in Python that performs the weighted sum and applies a step activation function for a perceptron. Test your function with sample weights, inputs, and bias values, and print the output for each case.<br><br><em><strong>Hint:</strong> Use NumPy arrays for vectorized operations.</em>

```python
import numpy as np

def perceptron(inputs, weights, bias):
    weighted_sum = np.dot(inputs, weights) + bias

    if weighted_sum >= 0:
        return 1
    else:
        return 0

# Test Cases
weights = np.array([0.5, 0.8])
bias = -0.6

inputs_list = [
    np.array([1, 1]),
    np.array([0, 1]),
    np.array([1, 0]),
    np.array([0, 0])
]

for inputs in inputs_list:
    output = perceptron(inputs, weights, bias)
    print("Input:", inputs, "Output:", output)
```

3. Demonstrate the limitation of a single-layer perceptron by attempting to train it (using your code) to solve the XOR problem (inputs: [0,0], [0,1], [1,0], [1,1]; outputs: 0, 1, 1, 0). Show and explain why it fails to classify correctly.

```python
import numpy as np

X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y = np.array([0,1,1,0])

weights = np.zeros(2)
bias = 0
learning_rate = 0.1

for epoch in range(20):

    for i in range(len(X)):

        prediction = 1 if np.dot(X[i],weights)+bias >=0 else 0

        error = y[i]-prediction

        weights += learning_rate*error*X[i]
        bias += learning_rate*error

print("Weights:",weights)
print("Bias:",bias)

print("\nPredictions")

for sample in X:

    prediction=1 if np.dot(sample,weights)+bias>=0 else 0

    print(sample,"Expected:",y[list(X).index(sample)],"Predicted:",prediction)
```

4. Implement a simple Multilayer Perceptron (MLP) in Python using only NumPy that can solve the XOR problem. Your network should have an input layer, one hidden layer, and an output layer, with manual forward propagation (no external libraries).<br><br><em><strong>Hint:</strong> Use sigmoid activation in the hidden and output layers.</em>

```python
import numpy as np

def sigmoid(x):
    return 1/(1+np.exp(-x))

# XOR Inputs
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

# Hidden layer weights
W1 = np.array([
    [20,-20],
    [20,-20]
])

b1 = np.array([-10,30])

# Output layer weights
W2 = np.array([
    [20],
    [20]
])

b2 = np.array([-30])

print("XOR Outputs\n")

for x in X:

    hidden = sigmoid(np.dot(x,W1)+b1)

    output = sigmoid(np.dot(hidden,W2)+b2)

    prediction = 1 if output>=0.5 else 0

    print("Input:",x,"Prediction:",prediction)
```

5. Modify your MLP code to classify whether a Zomato restaurant is 'trending' (1) or 'not trending' (0) based on two input features: average rating and number of recent reviews. Create a tiny dataset and run forward propagation to show the output for each sample.

```python
import numpy as np

def sigmoid(x):
    return 1/(1+np.exp(-x))

# Features:
# [Average Rating, Recent Reviews]

X = np.array([
    [4.8,500],
    [4.6,450],
    [3.2,80],
    [3.5,100]
])

# Random weights

W1 = np.array([
    [0.5,0.2],
    [0.01,0.03]
])

b1 = np.array([0.1,0.2])

W2 = np.array([
    [1.2],
    [1.1]
])

b2 = np.array([-1])

print("Trending Prediction\n")

for sample in X:

    hidden = sigmoid(np.dot(sample,W1)+b1)

    output = sigmoid(np.dot(hidden,W2)+b2)

    prediction = 1 if output>=0.5 else 0

    print("Input:",sample)
    print("Output:",round(float(output),3))
    print("Prediction:",prediction)
    print()
```
