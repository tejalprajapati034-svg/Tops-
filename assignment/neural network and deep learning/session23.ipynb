1. Load the ResNet50 model with pretrained ImageNet weights using Keras, print the model summary, and count how many layers are trainable by default.


```python
from tensorflow.keras.applications import ResNet50

# Load ResNet50
model = ResNet50(weights='imagenet')

# Print Model Summary
model.summary()

# Count Trainable Layers
trainable_layers = sum(layer.trainable for layer in model.layers)

print("Trainable Layers:", trainable_layers)
print("Total Layers:", len(model.layers))
```

```python
import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input
from tensorflow.keras.datasets import cifar10

# Load CIFAR-10
(X_train, y_train), (_, _) = cifar10.load_data()

# Select first 10 images
images = X_train[:10]

# Resize images to 224x224
images = tf.image.resize(images, (224, 224))

# Preprocess images
images = preprocess_input(images)

# Load VGG16 without top layer
model = VGG16(
    weights='imagenet',
    include_top=False
)

# Extract Features
features = model.predict(images)

print("Feature Shape:", features.shape)
```

3. Freeze all layers of the InceptionV3 model except the last 30 layers, then add a custom Dense output layer for 5-class classification, and print the number of trainable parameters.<br><br><em><strong>Hint:</strong> Use model.trainable = False, then set trainable=True for the last layers as needed.</em>

```python
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.layers import GlobalAveragePooling2D
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Model

# Load Base Model
base_model = InceptionV3(
    weights='imagenet',
    include_top=False
)

# Freeze All Layers
for layer in base_model.layers:
    layer.trainable = False

# Unfreeze Last 30 Layers
for layer in base_model.layers[-30:]:
    layer.trainable = True

# Add Custom Head
x = GlobalAveragePooling2D()(base_model.output)

output = Dense(
    5,
    activation='softmax'
)(x)

model = Model(
    inputs=base_model.input,
    outputs=output
)

# Count Trainable Parameters
print("Trainable Parameters:", model.count_params())
```
