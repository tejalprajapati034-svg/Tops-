1. Train a simple image classifier on the Fashion MNIST dataset using Keras and the SGD optimizer with default parameters, then record the accuracy and loss after 5 epochs.


```python
    import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD

# Load Dataset
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# Normalize
X_train = X_train / 255.0
X_test = X_test / 255.0

# Model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile
model.compile(
    optimizer=SGD(),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train
history = model.fit(
    X_train,
    y_train,
    epochs=5,
    validation_data=(X_test, y_test)
)

print("\nFinal Training Accuracy:", history.history['accuracy'][-1])
print("Final Training Loss:", history.history['loss'][-1])
print("Final Validation Accuracy:", history.history['val_accuracy'][-1])
print("Final Validation Loss:", history.history['val_loss'][-1])
```

3. Experiment with the Adam optimizer on the same Fashion MNIST model and adjust the learning rate to 0.0005; observe and note any changes in training speed or accuracy.<br><br><em><strong>Hint:</strong> In Keras, set learning_rate in the Adam optimizer: Adam(learning_rate=0.0005).</em>

```python
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import Adam

(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

X_train = X_train / 255.0
X_test = X_test / 255.0

model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(
    optimizer=Adam(learning_rate=0.0005),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    X_train,
    y_train,
    epochs=5,
    validation_data=(X_test,y_test)
)

print("\nValidation Accuracy:", history.history["val_accuracy"][-1])
```

4. Implement a learning rate schedule in your Keras model using the LearningRateScheduler callback to reduce the learning rate by half after every 2 epochs, and train with the Momentum optimizer; report how the learning rate changes during training.

```python
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.callbacks import LearningRateScheduler

(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

X_train = X_train / 255.0
X_test = X_test / 255.0

model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

optimizer = SGD(learning_rate=0.01, momentum=0.9)

model.compile(
    optimizer=optimizer,
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Learning Rate Scheduler
def scheduler(epoch, lr):
    if epoch > 0 and epoch % 2 == 0:
        lr = lr * 0.5
    print(f"Epoch {epoch+1}: Learning Rate = {lr}")
    return lr

lr_callback = LearningRateScheduler(scheduler)

history = model.fit(
    X_train,
    y_train,
    epochs=6,
    validation_data=(X_test, y_test),
    callbacks=[lr_callback]
)
```
