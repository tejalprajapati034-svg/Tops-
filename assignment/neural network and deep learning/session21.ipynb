1. Create a new Keras Sequential model and add a Conv2D layer with 32 filters, a 3x3 kernel, and ReLU activation for image input shape (32, 32, 3) as used in CIFAR-10.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D

# Create Model
model = Sequential()

# Add Conv2D Layer
model.add(
    Conv2D(
        filters=32,
        kernel_size=(3,3),
        activation='relu',
        input_shape=(32,32,3)
    )
)

# Display Model Summary
model.summary()
```

2. Add a MaxPooling2D layer after your Conv2D layer to reduce spatial dimensions, then flatten the output and add a Dense layer with 64 units and ReLU activation.

```python
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense

model = Sequential()

model.add(
    Conv2D(
        32,
        (3,3),
        activation='relu',
        input_shape=(32,32,3)
    )
)

model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())

model.add(Dense(64, activation='relu'))

model.summary()
```

3. Insert a BatchNormalization layer after the Conv2D and before MaxPooling2D, and add a Dropout layer (rate 0.3) after flattening to help prevent overfitting.<br><br><em><strong>Hint:</strong> BatchNorm stabilizes training, Dropout randomly disables neurons during training.</em>

```python
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import Dropout

model = Sequential()

model.add(
    Conv2D(
        32,
        (3,3),
        activation='relu',
        input_shape=(32,32,3)
    )
)

model.add(BatchNormalization())

model.add(MaxPooling2D((2,2)))

model.add(Flatten())

model.add(Dropout(0.3))

model.add(Dense(64, activation='relu'))

model.summary()
```

4. Compile your model using the 'adam' optimizer, 'categorical_crossentropy' loss, and track 'accuracy' as a metric. Then, train the model for 5 epochs using random data shaped like CIFAR-10 images and labels.<br><br><em><strong>Constraint:</strong> Use numpy to generate random input data and one-hot encoded labels for testing if you don't want to download the full CIFAR-10 dataset.</em>

```python
import numpy as np
from tensorflow.keras.utils import to_categorical

# Compile Model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Random CIFAR-10-like Images
X_train = np.random.rand(100,32,32,3)

# Random Labels
y = np.random.randint(0,10,100)

# One-Hot Encoding
y_train = to_categorical(y, num_classes=10)

# Add Output Layer
model.add(Dense(10, activation='softmax'))

# Recompile after adding the output layer
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Train Model
model.fit(
    X_train,
    y_train,
    epochs=5,
    batch_size=16
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Dense

model = Sequential()

# First Convolution Block
model.add(
    Conv2D(
        32,
        (3,3),
        activation='relu',
        input_shape=(32,32,3)
    )
)

model.add(BatchNormalization())

model.add(MaxPooling2D((2,2)))

# Second Convolution Block (Improvement)
model.add(
    Conv2D(
        64,
        (3,3),
        activation='relu'
    )
)

model.add(MaxPooling2D((2,2)))

# Fully Connected Layers
model.add(Flatten())

model.add(Dropout(0.3))

model.add(Dense(64, activation='relu'))

model.add(Dense(10, activation='softmax'))

model.summary()
```
