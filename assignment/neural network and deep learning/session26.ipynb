1. Prepare a sequence dataset from Flipkart product prices over 10 days (you can create fake data) and reshape it into the format required for LSTM input using NumPy.

```python
import numpy as np

# Fake Flipkart product prices (10 days)
prices = np.array([45000, 45200, 45150, 45500, 45800,
                   46000, 46250, 46500, 46700, 47000])

# Create sequences
X = []
y = []

sequence_length = 3

for i in range(len(prices) - sequence_length):
    X.append(prices[i:i + sequence_length])
    y.append(prices[i + sequence_length])

# Convert to NumPy arrays
X = np.array(X)
y = np.array(y)

# Reshape for LSTM
X = X.reshape((X.shape[0], X.shape[1], 1))

print("Input Shape:", X.shape)
print("Target Shape:", y.shape)

print("\nInput Sequences:")
print(X)

print("\nTarget Values:")
print(y)
```

2. Build a simple LSTM model in Keras to predict the next value in your prepared Flipkart price sequence and train it for 10 epochs.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Build LSTM Model
model = Sequential([
    LSTM(32, input_shape=(3,1)),
    Dense(1)
])

# Compile
model.compile(
    optimizer='adam',
    loss='mse'
)

# Train
model.fit(
    X,
    y,
    epochs=10,
    verbose=1
)
```

3. Evaluate your trained LSTM model by predicting the next price in the sequence, and print both the predicted and actual values for comparison.<br><br><em><strong>Hint:</strong> Use model.predict() and compare with the last actual value in your test set.</em>

```python
# Predict
prediction = model.predict(X)

print("Predicted Price:", prediction[-1][0])
print("Actual Price   :", y[-1])
```

4. Write a short code snippet to switch your LSTM layer to a GRU layer in the same model and retrain it. Compare the training time and prediction result with your LSTM model.<br><br><em><strong>Hint:</strong> Replace keras.layers.LSTM with keras.layers.GRU and keep other parameters the same.</em>

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense
import time

# Build GRU Model
gru_model = Sequential([
    GRU(32, input_shape=(3,1)),
    Dense(1)
])

gru_model.compile(
    optimizer='adam',
    loss='mse'
)

# Measure Training Time
start = time.time()

gru_model.fit(
    X,
    y,
    epochs=10,
    verbose=1
)

end = time.time()

print("Training Time:", end - start, "seconds")

# Prediction
prediction = gru_model.predict(X)

print("Predicted Price:", prediction[-1][0])
```

5. Use ChatGPT to generate a Python function that prepares any sequence data (such as daily Spotify song play counts) for LSTM input, and explain in your own words how the function works.

```python
import numpy as np

def prepare_sequence(data, sequence_length):

    X = []
    y = []

    for i in range(len(data) - sequence_length):
        X.append(data[i:i + sequence_length])
        y.append(data[i + sequence_length])

    X = np.array(X)
    y = np.array(y)

    X = X.reshape((X.shape[0], X.shape[1], 1))

    return X, y


# Example
spotify_plays = [120, 150, 180, 200, 210, 230, 250]

X, y = prepare_sequence(spotify_plays, 3)

print("Input Shape:", X.shape)
print("Target:", y)
```
