Prepare a sequence dataset using closing prices of any stock (download from Yahoo Finance or use a small CSV sample), and preprocess it into input sequences and targets suitable for training a GRU model in Keras.<br><br><em><strong>Hint:</strong> Use sliding windows to create input sequences of length 10 and the next value as the target.</em>


```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Load Dataset
df = pd.read_csv("stock_prices.csv")

prices = df["Close"].values.reshape(-1,1)

# Normalize Data
scaler = MinMaxScaler()

prices = scaler.fit_transform(prices)

# Create Sequences
sequence_length = 10

X = []
y = []

for i in range(len(prices)-sequence_length):

    X.append(prices[i:i+sequence_length])

    y.append(prices[i+sequence_length])

X = np.array(X)

y = np.array(y)

print("Input Shape:", X.shape)

print("Target Shape:", y.shape)
```

2. Build and train a simple GRU-based model (using Keras or PyTorch) to predict the next value in your prepared stock price sequence dataset. Print the training loss after each epoch.

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import MinMaxScaler

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU
from tensorflow.keras.layers import Dense

# Load Dataset
df = pd.read_csv("stock_prices.csv")

prices = df["Close"].values.reshape(-1,1)

scaler = MinMaxScaler()

prices = scaler.fit_transform(prices)

X=[]
y=[]

for i in range(len(prices)-10):

    X.append(prices[i:i+10])

    y.append(prices[i+10])

X=np.array(X)

y=np.array(y)

# Build Model
model=Sequential()

model.add(GRU(32,input_shape=(10,1)))

model.add(Dense(1))

model.compile(
    optimizer='adam',
    loss='mse'
)

# Train Model
history=model.fit(
    X,
    y,
    epochs=10,
    verbose=1
)

print("\nLoss Per Epoch")

for i,loss in enumerate(history.history["loss"]):

    print(f"Epoch {i+1}: {loss:.6f}")
```

3. Now build and train an LSTM-based model (same architecture as your GRU model except for the layer type) on the same dataset. Record the total training time and final accuracy for both GRU and LSTM models.<br><br><em><strong>Hint:</strong> Use Python's time module to measure training duration.</em>

```python
import time

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import GRU

# ---------- GRU ----------

gru=Sequential()

gru.add(GRU(32,input_shape=(10,1)))

gru.add(Dense(1))

gru.compile(
optimizer='adam',
loss='mse'
)

start=time.time()

gru_history=gru.fit(
X,
y,
epochs=10,
verbose=0
)

gru_time=time.time()-start

gru_loss=gru_history.history["loss"][-1]

# ---------- LSTM ----------

lstm=Sequential()

lstm.add(LSTM(32,input_shape=(10,1)))

lstm.add(Dense(1))

lstm.compile(
optimizer='adam',
loss='mse'
)

start=time.time()

lstm_history=lstm.fit(
X,
y,
epochs=10,
verbose=0
)

lstm_time=time.time()-start

lstm_loss=lstm_history.history["loss"][-1]

print("GRU Time :",round(gru_time,2),"seconds")
print("GRU Loss :",gru_loss)

print()

print("LSTM Time :",round(lstm_time,2),"seconds")
print("LSTM Loss :",lstm_loss)
```
