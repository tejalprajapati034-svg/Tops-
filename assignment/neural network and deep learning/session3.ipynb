1. Create a simple Python function called neuron_output that takes three arguments: input_value, weight, and bias, and returns the output using the formula output = input_value * weight + bias.



```python
def neuron_output(input_value, weight, bias):
    output = input_value * weight + bias
    return output

# Example
result = neuron_output(10, 0.8, 2)

print("Output:", result)
```

2. Given an array of input values representing the number of likes on 5 Instagram posts, apply a weight of 0.8 and a bias of 2 to each input using your neuron_output function, and print the resulting outputs.<br><br><em><strong>Hint:</strong> Use a for loop to process each input value.</em>

```python
def neuron_output(input_value, weight, bias):
    return input_value * weight + bias

# Number of likes on 5 Instagram posts
likes = [50, 120, 80, 200, 150]

weight = 0.8
bias = 2

print("Neuron Outputs:\n")

for like in likes:
    output = neuron_output(like, weight, bias)
    print(f"Likes = {like} --> Output = {output}")
```

3. Write Python code to initialize a list of 10 neuron weights using three different strategies: all zeros, random values between -1 and 1, and Xavier initialization (drawn from a normal distribution with mean 0 and standard deviation 1/sqrt(n), where n=10). Print the three lists for comparison.<br><br><em><strong>Hint:</strong> Use the numpy library for random and normal distributions.</em>

```python
import numpy as np

n = 10

# 1. All Zeros
zeros = np.zeros(n)

# 2. Random values between -1 and 1
random_weights = np.random.uniform(-1, 1, n)

# 3. Xavier Initialization
xavier = np.random.normal(0, 1/np.sqrt(n), n)

print("Zero Initialization:")
print(zeros)

print("\nRandom Initialization:")
print(random_weights)

print("\nXavier Initialization:")
print(xavier)
```


4. Demonstrate with code why having a bias term is essential: set up a neuron with weight=0.5 and bias=0, and another with weight=0.5 and bias=5. For an input of 0, show the outputs for both cases and explain the difference in a comment.

```python
def neuron_output(input_value, weight, bias):
    return input_value * weight + bias

weight = 0.5
input_value = 0

output1 = neuron_output(input_value, weight, 0)
output2 = neuron_output(input_value, weight, 5)

print("Bias = 0 --> Output =", output1)
print("Bias = 5 --> Output =", output2)

# Bias shifts the output even when the input is zero.
# Without bias, the output remains 0 for an input of 0.
```

5. Modify the neuron_output function to accept a list of input values and a corresponding list of weights, plus a single bias. Compute the weighted sum plus bias for a Spotify playlist (inputs: [2, 4, 1], weights: [0.6, 0.3, 0.1], bias: 1) and print the result.<br><br><em><strong>Constraint:</strong> Do not use any external libraries except basic Python (no numpy).</em>

```python
def neuron_output(inputs, weights, bias):
    weighted_sum = 0

    for i in range(len(inputs)):
        weighted_sum += inputs[i] * weights[i]

    return weighted_sum + bias


# Spotify Playlist Example
inputs = [2, 4, 1]
weights = [0.6, 0.3, 0.1]
bias = 1

result = neuron_output(inputs, weights, bias)

print("Weighted Sum + Bias =", result)
```
