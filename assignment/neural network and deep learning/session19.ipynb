1. Create a 5x5 NumPy matrix with random integers from 0 to 9, then define a 3x3 filter (kernel) and manually perform a single convolution operation (no padding, stride 1) to produce the activation map.

```python
import numpy as np

# Create 5x5 input matrix
input_matrix = np.random.randint(0, 10, (5, 5))

print("Input Matrix:")
print(input_matrix)

# 3x3 Kernel
kernel = np.array([
    [1, 0, -1],
    [1, 0, -1],
    [1, 0, -1]
])

print("\nKernel:")
print(kernel)

# Output Activation Map (3x3)
output = np.zeros((3, 3))

for i in range(3):
    for j in range(3):
        region = input_matrix[i:i+3, j:j+3]
        output[i, j] = np.sum(region * kernel)

print("\nActivation Map:")
print(output)
```

2. Implement a Python function apply_stride_convolution(input_matrix, kernel, stride) that performs convolution with a given stride on a 6x6 matrix and a 3x3 kernel, returning the resulting activation map.<br><br><em><strong>Hint:</strong> Use nested loops to move the kernel by the stride value over the input matrix.</em>

```python
import numpy as np

def apply_stride_convolution(input_matrix, kernel, stride):

    n = input_matrix.shape[0]
    k = kernel.shape[0]

    output_size = ((n - k) // stride) + 1

    output = np.zeros((output_size, output_size))

    for i in range(0, n-k+1, stride):
        for j in range(0, n-k+1, stride):

            region = input_matrix[i:i+k, j:j+k]

            output[i//stride, j//stride] = np.sum(region * kernel)

    return output


# Example
matrix = np.random.randint(0, 10, (6,6))

kernel = np.array([
    [1,0,-1],
    [1,0,-1],
    [1,0,-1]
])

result = apply_stride_convolution(matrix, kernel, stride=2)

print(result)
```

3. Modify your convolution code to add zero-padding of 1 around the input matrix before applying the convolution, and show the new activation map output.

```python
import numpy as np

# Original Matrix
matrix = np.random.randint(0,10,(5,5))

# Add Zero Padding
padded = np.pad(
    matrix,
    pad_width=1,
    mode='constant',
    constant_values=0
)

kernel = np.array([
    [1,0,-1],
    [1,0,-1],
    [1,0,-1]
])

output = np.zeros((5,5))

for i in range(5):
    for j in range(5):

        region = padded[i:i+3, j:j+3]

        output[i,j] = np.sum(region * kernel)

print("Original Matrix:")
print(matrix)

print("\nPadded Matrix:")
print(padded)

print("\nActivation Map:")
print(output)
```

4. Visualize the effect of different filters by applying both an edge detection kernel and a blur kernel to the same 5x5 matrix and print both activation maps side by side.<br><br><em><strong>Hint:</strong> Use the Sobel or Laplacian kernel for edge detection and an averaging kernel for blur.</em>

```python
import numpy as np

matrix = np.random.randint(0,10,(5,5))

# Edge Detection Kernel (Sobel-like)
edge_kernel = np.array([
    [1,0,-1],
    [2,0,-2],
    [1,0,-1]
])

# Blur Kernel
blur_kernel = np.ones((3,3)) / 9

def convolution(matrix, kernel):

    output = np.zeros((3,3))

    for i in range(3):
        for j in range(3):

            region = matrix[i:i+3,j:j+3]

            output[i,j] = np.sum(region * kernel)

    return output

edge_output = convolution(matrix, edge_kernel)

blur_output = convolution(matrix, blur_kernel)

print("Input Matrix:")
print(matrix)

print("\nEdge Detection Output:")
print(edge_output)

print("\nBlur Output:")
print(blur_output)
```
