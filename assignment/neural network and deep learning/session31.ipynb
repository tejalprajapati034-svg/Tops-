1. Save your trained music genre classification model (for example, a model that predicts if a song is Pop, Rock, or Classical based on audio features) to a file using joblib or pickle.


```python
import pandas as pd
import numpy as np
import joblib

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier

# -----------------------
# Create Dummy Dataset
# -----------------------

np.random.seed(42)

data = pd.DataFrame({
    "tempo": np.random.randint(60,180,200),
    "energy": np.random.rand(200),
    "danceability": np.random.rand(200),
    "genre": np.random.choice(
        ["Pop","Rock","Classical"],
        200
    )
})

X = data[["tempo","energy","danceability"]]
y = data["genre"]

# -----------------------
# Scale Data
# -----------------------

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

# -----------------------
# Train Model
# -----------------------

X_train,X_test,y_train,y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)

model = RandomForestClassifier()

model.fit(X_train,y_train)

print("Accuracy:",model.score(X_test,y_test))

# Save Model
joblib.dump(model,"genre_model.pkl")
joblib.dump(scaler,"scaler.pkl")

print("Model Saved Successfully")
```

2. Build an inference pipeline in Python that loads your saved music genre model and predicts the genre for a new set of audio features provided as input.<br><br><em><strong>Hint:</strong> The pipeline should include data preprocessing steps if required, just like during training.</em>

```python
import joblib
import numpy as np

model = joblib.load("genre_model.pkl")
scaler = joblib.load("scaler.pkl")

# New Song

new_song = np.array([[120,0.80,0.70]])

new_song = scaler.transform(new_song)

prediction = model.predict(new_song)

print("Predicted Genre:",prediction[0])
```

3. Create a Streamlit upload interface that lets users upload a CSV file containing audio features of multiple songs, then displays the predicted genre for each song in a table

```python
import streamlit as st
import pandas as pd
import joblib

model = joblib.load("genre_model.pkl")
scaler = joblib.load("scaler.pkl")

st.title("🎵 Music Genre Classifier")

uploaded_file = st.file_uploader(
    "Upload CSV File",
    type=["csv"]
)

if uploaded_file:

    df = pd.read_csv(uploaded_file)

    X = scaler.transform(df)

    prediction = model.predict(X)

    df["Predicted Genre"] = prediction

    st.write(df)
```

4. Add real-time prediction to your Streamlit app: as soon as a user uploads a single song's features, instantly display the predicted genre with a confidence score, similar to how Spotify recommends genres.

```python
import streamlit as st
import numpy as np
import joblib

model = joblib.load("genre_model.pkl")
scaler = joblib.load("scaler.pkl")

st.title("🎵 Spotify Genre Predictor")

tempo = st.slider("Tempo",60,180,120)

energy = st.slider("Energy",0.0,1.0,0.5)

dance = st.slider("Danceability",0.0,1.0,0.5)

if st.button("Predict"):

    X = np.array([[tempo,energy,dance]])

    X = scaler.transform(X)

    prediction = model.predict(X)[0]

    confidence = max(model.predict_proba(X)[0])

    st.success(f"Genre : {prediction}")

    st.write(f"Confidence : {confidence:.2%}")
```
