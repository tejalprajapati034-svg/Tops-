
1. Implement the Leaky ReLU activation function in Python as a function leaky_relu(x, alpha=0.01), and test it on an array of values ranging from -5 to 5.


```python
import numpy as np

# Leaky ReLU Function
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

# Test values
x = np.arange(-5, 6)

output = leaky_relu(x)

print("Input :", x)
print("Output:", output)
```

2. Write a Python function parametric_relu(x, a) where 'a' is a learnable parameter (try a=0.2), and plot its output for x values from -5 to 5 using matplotlib.<br><br><em><strong>Hint:</strong> Compare the graph with standard ReLU and Leaky ReLU on the same plot for better understanding.</em>

```python
import numpy as np
import matplotlib.pyplot as plt

# ReLU
def relu(x):
    return np.maximum(0, x)

# Leaky ReLU
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

# Parametric ReLU
def parametric_relu(x, a):
    return np.where(x > 0, x, a * x)

x = np.linspace(-5, 5, 200)

plt.plot(x, relu(x), label="ReLU")
plt.plot(x, leaky_relu(x), label="Leaky ReLU")
plt.plot(x, parametric_relu(x, 0.2), label="PReLU (a=0.2)")

plt.title("ReLU vs Leaky ReLU vs PReLU")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.grid(True)

plt.show()
```

3. Given a list of predicted scores for 4 classes: [2.0, 1.0, 0.1, 0.5], implement the softmax function in Python and print the resulting probabilities rounded to 3 decimal places.

```python
import numpy as np

def softmax(scores):
    exp_scores = np.exp(scores)
    return exp_scores / np.sum(exp_scores)

scores = np.array([2.0, 1.0, 0.1, 0.5])

probabilities = softmax(scores)

print("Softmax Probabilities:")

for p in probabilities:
    print(round(p, 3))
```

4. Write a function hard_sigmoid(x) that implements the hard sigmoid activation (output between 0 and 1), and test it on inputs: -3, -1, 0, 1, 3.<br><br><em><strong>Hint:</strong> Use the formula: hard_sigmoid(x) = max(0, min(1, 0.2*x + 0.5)).</em>

```python
def hard_sigmoid(x):
    return max(0, min(1, 0.2 * x + 0.5))

inputs = [-3, -1, 0, 1, 3]

for value in inputs:
    print(f"Input: {value}, Output: {hard_sigmoid(value)}")
```

5. Suppose you are building a multi-class image classifier for food items (like Zomato's food image feed). For each layer below, choose and justify the best activation function: (a) hidden dense layer, (b) output layer with 5 classes, (c) output layer with binary classification.<br><br><em><strong>Constraint:</strong> Write 1-2 lines of reasoning for each choice.</em>

#### (a) Hidden Dense Layer

#### Activation: ReLU

- Reason: ReLU is fast, computationally efficient, and helps reduce the vanishing gradient problem, making it the standard choice for hidden layers.

#### (b) Output Layer (5 Classes)

#### Activation: Softmax

- Reason: Softmax converts the output scores into probabilities that sum to 1, making it ideal for multi-class classification (5 food categories).

#### (c) Output Layer (Binary Classification)

- Activation: Sigmoid

- Reason: Sigmoid produces a value between 0 and 1, which can be interpreted as the probability of belonging to the positive class, making it suitable for binary classification.
