2. Create a Sequential model in Keras with three layers: an input Dense layer with 32 units and 'relu' activation, a Dense layer with 16 units and 'relu', and an output Dense layer with 1 unit and 'sigmoid'. Print the model summary.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Create Model
model = Sequential([
    Dense(32, activation="relu", input_shape=(10,)),
    Dense(16, activation="relu"),
    Dense(1, activation="sigmoid")
])

# Print Summary
model.summary()
```

3. Build a simple Keras Functional API model that takes a 10-feature input, passes it through one Dense layer with 20 units and 'relu', then outputs a single value with a Dense layer. Display the model summary and total parameter count.

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model

# Input Layer
inputs = Input(shape=(10,))

# Hidden Layer
hidden = Dense(20, activation="relu")(inputs)

# Output Layer
outputs = Dense(1)(hidden)

# Create Model
model = Model(inputs=inputs, outputs=outputs)

# Print Summary
model.summary()

print("\nTotal Parameters:", model.count_params())
```

4. Write code to check if TensorFlow is using your CPU or GPU for computations. Print out the list of available devices and indicate which one is being used.<br><br><em><strong>Hint:</strong> Use tf.config.list_physical_devices() and tf.test.is_gpu_available().</em>

```python
import tensorflow as tf

print("TensorFlow Version:", tf.__version__)

# CPU Devices
cpu = tf.config.list_physical_devices("CPU")
print("\nCPU Devices:")
print(cpu)

# GPU Devices
gpu = tf.config.list_physical_devices("GPU")
print("\nGPU Devices:")
print(gpu)

# Check GPU Availability
if len(gpu) > 0:
    print("\nTensorFlow is using GPU.")
else:
    print("\nTensorFlow is using CPU.")
```

Use ChatGPT or Copilot to generate Keras code for a model with two inputs (e.g., one for user age and one for number of songs listened on Spotify) and one output (predicting likelihood of buying a premium subscription). Paste the generated code, run it, and note any errors or changes you had to make.

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Concatenate
from tensorflow.keras.models import Model

# Input 1: User Age
age_input = Input(shape=(1,), name="Age")

# Input 2: Songs Listened
songs_input = Input(shape=(1,), name="Songs_Listened")

# Combine Inputs
merged = Concatenate()([age_input, songs_input])

# Hidden Layer
hidden = Dense(16, activation="relu")(merged)

# Output Layer
output = Dense(1, activation="sigmoid")(hidden)

# Model
model = Model(
    inputs=[age_input, songs_input],
    outputs=output
)

# Summary
model.summary()
```
