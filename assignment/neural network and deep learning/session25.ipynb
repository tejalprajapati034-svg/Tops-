1. Create a simple Python script using Keras to build an LSTM layer for predicting the next value in a sequence of daily step counts (e.g., [4000, 4200, 4300, 4100, 4500, ...]) like those tracked by a fitness app.


```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Daily step counts
steps = np.array([
    4000, 4200, 4300, 4100,
    4500, 4700, 4900, 5000
])

# Create sequences
X = []
y = []

for i in range(len(steps)-3):
    X.append(steps[i:i+3])
    y.append(steps[i+3])

X = np.array(X)
y = np.array(y)

# Reshape for LSTM
X = X.reshape((X.shape[0], X.shape[1], 1))

# Build Model
model = Sequential([
    LSTM(32, input_shape=(3,1)),
    Dense(1)
])

# Compile
model.compile(
    optimizer='adam',
    loss='mse'
)

# Train
model.fit(
    X,
    y,
    epochs=20,
    verbose=1
)
```

2. Write a short code snippet to demonstrate the vanishing gradient problem in a basic RNN by training it on a sequence of 0s and 1s, then print the gradients after each epoch.<br><br><em><strong>Hint:</strong> Use TensorFlow or PyTorch and compare the gradients at early vs late time steps.</em>

```python
import tensorflow as tf
import numpy as np

# Dummy sequence
X = tf.constant(np.random.randint(0,2,(1,10,1)), dtype=tf.float32)
y = tf.constant([[1.0]])

# Simple RNN Model
model = tf.keras.Sequential([
    tf.keras.layers.SimpleRNN(8),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

optimizer = tf.keras.optimizers.Adam()

# Training Loop
for epoch in range(5):

    with tf.GradientTape() as tape:

        pred = model(X)

        loss = tf.keras.losses.binary_crossentropy(y, pred)

    gradients = tape.gradient(
        loss,
        model.trainable_variables
    )

    optimizer.apply_gradients(
        zip(gradients, model.trainable_variables)
    )

    print(f"\nEpoch {epoch+1}")

    for g in gradients:
        print(tf.reduce_mean(tf.abs(g)).numpy())
```

#### Gate	
- Forget Gate :	Removes unnecessary information from the previous cell state.
- Input Gate	: Decides which new information should be stored.
- Cell State :	Stores long-term information across time steps.
- Output Gate : Decides what information is passed to the next hidden state.

5. Write a short explanation (3-4 lines) of what 'unrolling an LSTM through time' means, using an example from a music playlist app that predicts the next song a user might play.

#### Answer:

- Unrolling an LSTM through time means applying the same LSTM cell repeatedly to each element of a sequence while passing the hidden state and cell state from one time step to the next. In a music playlist app, the LSTM processes each song the user listens to in order and remembers previous listening patterns. Using this information, it predicts the next song the user is most likely to play.
