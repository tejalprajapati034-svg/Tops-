1. Write a Python function calculate_mse(predictions, targets) that takes two lists of numbers and returns the Mean Squared Error between them using the MSE formula.


```python
def calculate_mse(predictions, targets):
    n = len(predictions)

    mse = 0

    for i in range(n):
        mse += (predictions[i] - targets[i]) ** 2

    mse = mse / n

    return mse


# Example
predictions = [4, 5, 3, 2]
targets = [5, 4, 3, 1]

print("Mean Squared Error:", calculate_mse(predictions, targets))
```

2. Given two arrays: actual_ratings = [4, 3, 5, 2, 1] and predicted_ratings = [5, 2, 4, 2, 1], calculate the Mean Absolute Error (MAE) using Python and print the result.

```python
actual_ratings = [4, 3, 5, 2, 1]
predicted_ratings = [5, 2, 4, 2, 1]

n = len(actual_ratings)

mae = 0

for i in range(n):
    mae += abs(actual_ratings[i] - predicted_ratings[i])

mae = mae / n

print("Mean Absolute Error:", mae)
```

3. Implement a function binary_cross_entropy(y_true, y_pred) in Python that computes the binary cross-entropy loss for a batch of predictions and true labels (both as lists of 0s and 1s), then test it with y_true = [1, 0, 1] and y_pred = [0.9, 0.2, 0.7].

```python
import math

def binary_cross_entropy(y_true, y_pred):

    n = len(y_true)
    loss = 0

    for i in range(n):
        loss += -(y_true[i] * math.log(y_pred[i]) +
                  (1 - y_true[i]) * math.log(1 - y_pred[i]))

    return loss / n


y_true = [1, 0, 1]
y_pred = [0.9, 0.2, 0.7]

print("Binary Cross Entropy Loss:",
      round(binary_cross_entropy(y_true, y_pred), 4))
```

4. Simulate a basic Categorical Cross-Entropy calculation: given y_true = [0, 0, 1] and y_pred = [0.2, 0.3, 0.5], write Python code to compute the categorical cross-entropy loss for this single data point.

```python
import math

y_true = [0, 0, 1]
y_pred = [0.2, 0.3, 0.5]

loss = 0

for i in range(len(y_true)):
    if y_true[i] == 1:
        loss = -math.log(y_pred[i])

print("Categorical Cross Entropy Loss:",
      round(loss, 4))
```
