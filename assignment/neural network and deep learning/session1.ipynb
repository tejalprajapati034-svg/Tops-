1. Open TensorFlow Playground (https://playground.tensorflow.org/) and create a simple neural network with 1 hidden layer and 3 neurons to classify the orange and blue dots. Change the learning rate and observe how the decision boundary changes.<br><br><em><strong>Hint:</strong> Try both a low and high learning rate to see which one converges faster.</em>

#### Steps:

- Open https://playground.tensorflow.org/
- Select Classification dataset (orange and blue dots).
#### Set:
- Hidden Layers = 1
- Neurons = 3
#### Try two learning rates:
- Low Learning Rate = 0.01
- High Learning Rate = 1.0
- Click Run and observe the decision boundary.
#### Observation
- Low Learning Rate (0.01):
- The model learns slowly.
- The decision boundary changes gradually.
- It usually converges more accurately but takes more time.
#### High Learning Rate (1.0):
- The model learns very quickly.
- The decision boundary changes rapidly.
- It may overshoot the best solution and produce unstable results.
