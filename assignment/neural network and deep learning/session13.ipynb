1. Install TensorFlow and Keras in your Python environment, then create a new file called spotify_like_classifier.py.



```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping

# -----------------------------
# Generate Random Dataset
# -----------------------------

np.random.seed(42)

# 100 user profiles
# Feature 1: Listening Time (minutes)
listening_time = np.random.randint(10, 300, 100)

# Feature 2: Liked Genre Score (0 to 1)
liked_genre_score = np.random.rand(100)

# Combine features
X = np.column_stack((listening_time, liked_genre_score))

# Binary Labels (0 or 1)
y = np.random.randint(0, 2, 100)

# Normalize listening time
X[:, 0] = X[:, 0] / 300.0

# -----------------------------
# Create Sequential Model
# -----------------------------

model = Sequential([
    Dense(16, activation='relu', input_shape=(2,)),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')
])

# -----------------------------
# Compile Model
# -----------------------------

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# -----------------------------
# Early Stopping Callback
# -----------------------------

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
    
)

# -----------------------------
# Train Model
# -----------------------------

history = model.fit(
    X,
    y,
    epochs=20,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)

# -----------------------------
# Final Accuracy
# -----------------------------

print("\nFinal Training Accuracy:",
      round(history.history["accuracy"][-1], 4))

print("Final Validation Accuracy:",
      round(history.history["val_accuracy"][-1], 4))
```

2. In spotify_like_classifier.py, define a Sequential model in Keras with three Dense layers: the first with 16 units and 'relu' activation, the second with 8 units and 'relu', and the third with 1 unit and 'sigmoid' activation.\

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([
    Dense(16, activation='relu', input_shape=(2,)),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

3. Compile your model using the 'adam' optimizer, 'binary_crossentropy' loss, and include 'accuracy' as a metric.

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
import numpy as np
import tensorflow as tf

np.random.seed(42)
tf.random.set_seed(42)

# Generate Features
listening_time = np.random.rand(100, 1) * 100
liked_genre_score = np.random.rand(100, 1)

# Combine Features
X = np.hstack((listening_time, liked_genre_score))

# Generate Labels
y = np.random.randint(0, 2, size=(100, 1))

# Train Model
history = model.fit(
    X,
    y,
    epochs=20,
    validation_split=0.2,
    verbose=1
)

# Evaluate Model
loss, accuracy = model.evaluate(X, y, verbose=0)

print("Accuracy:", accuracy)
```
