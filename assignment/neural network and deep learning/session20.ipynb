1. Create a 6x6 NumPy array with random integers (0-255) to simulate a grayscale image, then apply a 2x2 max pooling operation with a stride of 2 and print the resulting reduced array.

```python
import numpy as np

# Create 6x6 random grayscale image
image = np.random.randint(0, 256, (6, 6))

print("Original Image:")
print(image)

# Max Pooling (2x2, stride=2)
output = np.zeros((3, 3), dtype=int)

for i in range(0, 6, 2):
    for j in range(0, 6, 2):
        output[i//2, j//2] = np.max(image[i:i+2, j:j+2])

print("\nMax Pooled Image:")
print(output)
```

2. Write a Python function average_pooling_2x2(image) that takes a 2D NumPy array and returns a new array after applying 2x2 average pooling with stride 2.

```python
import numpy as np

def average_pooling_2x2(image):

    rows, cols = image.shape

    output = np.zeros((rows//2, cols//2))

    for i in range(0, rows, 2):
        for j in range(0, cols, 2):

            block = image[i:i+2, j:j+2]
            output[i//2, j//2] = np.mean(block)

    return output


# Example
image = np.array([
    [2,4,6,8],
    [1,3,5,7],
    [8,6,4,2],
    [7,5,3,1]
])

print(average_pooling_2x2(image))
```

4. Combine a single 3x3 convolution layer (use any simple filter, e.g., edge detection) followed by a 2x2 max pooling layer on a 5x5 input array. Show the output after each step.

```python
import numpy as np

# Input Image
image = np.array([
    [1,2,3,4,5],
    [5,6,7,8,9],
    [9,8,7,6,5],
    [5,4,3,2,1],
    [1,2,3,4,5]
])

# Edge Detection Filter
kernel = np.array([
    [1,0,-1],
    [1,0,-1],
    [1,0,-1]
])

# Convolution
conv_output = np.zeros((3,3))

for i in range(3):
    for j in range(3):
        region = image[i:i+3, j:j+3]
        conv_output[i,j] = np.sum(region * kernel)

print("Convolution Output:")
print(conv_output)

# Max Pooling
pool_output = np.zeros((1,1))

pool_output[0,0] = np.max(conv_output[0:2,0:2])

print("\nMax Pooling Output:")
print(pool_output)
```
