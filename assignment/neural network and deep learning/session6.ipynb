1. Create a Python script using NumPy to perform a manual forward pass for a single-layer neural network: given an input vector X and a weight matrix W, compute the dot product and apply the sigmoid activation function to get the output.


```python
import numpy as np

# Sigmoid Activation Function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Input Vector
X = np.array([1.0, 2.0, 3.0])

# Weight Matrix (3 inputs → 2 neurons)
W = np.array([
    [0.2, 0.4],
    [0.3, 0.5],
    [0.6, 0.1]
])

# Forward Pass
z = np.dot(X, W)
output = sigmoid(z)

print("Input:")
print(X)

print("\nWeighted Sum:")
print(z)

print("\nOutput after Sigmoid:")
print(output)
```

2. Modify your script to handle a two-layer neural network: take an input vector, compute the first hidden layer's activations using a weight matrix and ReLU activation, then compute the output layer's activations using a second weight matrix and sigmoid activation.

```python
import numpy as np

# Activation Functions
def relu(x):
    return np.maximum(0, x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Input
X = np.array([1.0, 2.0, 3.0])

# Input → Hidden Weights (3×3)
W1 = np.array([
    [0.2, 0.4, 0.1],
    [0.5, 0.3, 0.2],
    [0.6, 0.1, 0.4]
])

# Hidden → Output Weights (3×2)
W2 = np.array([
    [0.2, 0.5],
    [0.3, 0.4],
    [0.6, 0.1]
])

# Hidden Layer
hidden_input = np.dot(X, W1)
hidden_output = relu(hidden_input)

# Output Layer
output_input = np.dot(hidden_output, W2)
output = sigmoid(output_input)

print("Hidden Layer Activations:")
print(hidden_output)

print("\nOutput Layer:")
print(output)
```

3. Visualize the activations of each layer in your two-layer network using matplotlib — plot the input, hidden, and output activations as bar charts to show how data transforms through the network.<br><br><em><strong>Hint:</strong> Use plt.bar() for each layer and display them in separate subplots.</em>

```python
import numpy as np
import matplotlib.pyplot as plt

# Activation Functions
def relu(x):
    return np.maximum(0, x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Input
X = np.array([1.0, 2.0, 3.0])

# Weights
W1 = np.array([
    [0.2, 0.4, 0.1],
    [0.5, 0.3, 0.2],
    [0.6, 0.1, 0.4]
])

W2 = np.array([
    [0.2, 0.5],
    [0.3, 0.4],
    [0.6, 0.1]
])

# Forward Pass
hidden = relu(np.dot(X, W1))
output = sigmoid(np.dot(hidden, W2))

# Plot
plt.figure(figsize=(10, 4))

plt.subplot(1, 3, 1)
plt.bar(range(len(X)), X)
plt.title("Input")

plt.subplot(1, 3, 2)
plt.bar(range(len(hidden)), hidden)
plt.title("Hidden Layer")

plt.subplot(1, 3, 3)
plt.bar(range(len(output)), output)
plt.title("Output Layer")

plt.tight_layout()
plt.show()
```

4. Given a set of random weights and a batch of three input vectors, perform a forward pass through a three-layer network (input, one hidden, output) and print the output activations for each input.<br><br><em><strong>Constraint:</strong> Use only NumPy for all matrix operations and activations.</em>

```python
import numpy as np

# Activation Functions
def relu(x):
    return np.maximum(0, x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Batch of 3 input vectors
X = np.array([
    [1, 2, 3],
    [2, 1, 0],
    [3, 2, 1]
])

# Random seed (for reproducibility)
np.random.seed(42)

# Random Weights
W1 = np.random.rand(3, 4)   # Input → Hidden
W2 = np.random.rand(4, 2)   # Hidden → Output

# Forward Pass
hidden = relu(np.dot(X, W1))
output = sigmoid(np.dot(hidden, W2))

print("Input Batch:")
print(X)

print("\nHidden Layer Activations:")
print(hidden)

print("\nOutput Activations:")
print(output)
```
