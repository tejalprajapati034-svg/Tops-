1. Create a simple Keras Sequential model for classifying handwritten digits (use the MNIST dataset), and train it without any regularization. Save the validation accuracy after 5 epochs.


```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

# Load Dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize
X_train = X_train / 255.0
X_test = X_test / 255.0

# Model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    X_train,
    y_train,
    epochs=5,
    validation_data=(X_test, y_test)
)

print("\nValidation Accuracy after 5 epochs:")
print(history.history['val_accuracy'][-1])
```

2. Modify your Keras model to add L2 regularization (kernel_regularizer=regularizers.l2(0.01)) to all Dense layers, retrain for 5 epochs, and compare the validation accuracy to your previous result.

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras import regularizers

(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train = X_train / 255.0
X_test = X_test / 255.0

model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128,
          activation='relu',
          kernel_regularizer=regularizers.l2(0.01)),
    Dense(64,
          activation='relu',
          kernel_regularizer=regularizers.l2(0.01)),
    Dense(10,
          activation='softmax',
          kernel_regularizer=regularizers.l2(0.01))
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    X_train,
    y_train,
    epochs=5,
    validation_data=(X_test,y_test)
)

print("Validation Accuracy:")
print(history.history["val_accuracy"][-1])
```

3. Add Dropout layers (with rate=0.5) after each Dense layer in your Keras model, retrain for 5 epochs, and observe the change in validation accuracy.<br><br><em><strong>Hint:</strong> Use keras.layers.Dropout and place it after activations.</em>

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout

(X_train,y_train),(X_test,y_test)=mnist.load_data()

X_train=X_train/255.0
X_test=X_test/255.0

model=Sequential([
    Flatten(input_shape=(28,28)),

    Dense(128,activation='relu'),
    Dropout(0.5),

    Dense(64,activation='relu'),
    Dropout(0.5),

    Dense(10,activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history=model.fit(
    X_train,
    y_train,
    epochs=5,
    validation_data=(X_test,y_test)
)

print("Validation Accuracy:")
print(history.history["val_accuracy"][-1])
```

4. Integrate BatchNormalization layers into your Keras model (before activation functions in Dense layers), retrain, and note how the training and validation accuracy curves change.<br><br><em><strong>Hint:</strong> Use keras.layers.BatchNormalization before the activation in each Dense block.</em>

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, BatchNormalization, Activation

(X_train,y_train),(X_test,y_test)=mnist.load_data()

X_train=X_train/255.0
X_test=X_test/255.0

model=Sequential([
    Flatten(input_shape=(28,28)),

    Dense(128),
    BatchNormalization(),
    Activation("relu"),

    Dense(64),
    BatchNormalization(),
    Activation("relu"),

    Dense(10,activation="softmax")
])

model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

history=model.fit(
    X_train,
    y_train,
    epochs=5,
    validation_data=(X_test,y_test)
)

print("Training Accuracy:")
print(history.history["accuracy"])

print("\nValidation Accuracy:")
print(history.history["val_accuracy"])
```

5. Implement EarlyStopping in your model training using keras.callbacks.EarlyStopping with patience=2, and show how it affects the number of training epochs and final validation accuracy.

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.callbacks import EarlyStopping

(X_train,y_train),(X_test,y_test)=mnist.load_data()

X_train=X_train/255.0
X_test=X_test/255.0

model=Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128,activation="relu"),
    Dense(64,activation="relu"),
    Dense(10,activation="softmax")
])

model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

early_stop=EarlyStopping(
    monitor="val_loss",
    patience=2,
    restore_best_weights=True
)

history=model.fit(
    X_train,
    y_train,
    epochs=20,
    validation_data=(X_test,y_test),
    callbacks=[early_stop]
)

print("\nTraining stopped after", len(history.history["loss"]), "epochs")
print("Final Validation Accuracy:", history.history["val_accuracy"][-1])
```
