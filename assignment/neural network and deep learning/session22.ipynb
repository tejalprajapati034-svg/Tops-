1. Load any 3 sample images (your choice: selfies, food, cricket, or memes) and use OpenCV or PIL to resize each to 224x224 pixels, saving the resized versions in a new folder.


```python
from PIL import Image
import os

# Create output folder
os.makedirs("resized_images", exist_ok=True)

# List of images
images = [
    "virat_kohli.jpg",
    "virat_kohli.jpg",
    "virat_kohli.jpg"
]

for img_name in images:
    img = Image.open(img_name)

    # Resize image
    resized = img.resize((224, 224))

    # Save resized image
    resized.save("resized_images/" + img_name)

print("All images resized successfully!")
```

2. Convert a colored image (choose any photo you like) to grayscale and then back to RGB using Python libraries. Display all three versions (original, grayscale, RGB) using matplotlib.

```python
from PIL import Image
import matplotlib.pyplot as plt

# Load Image
image = Image.open("virat_kohli.jpg")

# Convert to Grayscale
gray = image.convert("L")

# Convert back to RGB
rgb = gray.convert("RGB")

# Display Images
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(image)
plt.title("Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(gray, cmap="gray")
plt.title("Grayscale")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(rgb)
plt.title("RGB")
plt.axis("off")

plt.show()
```

3. Use Keras' ImageDataGenerator to apply random rotation and horizontal flip to an image of your choice (for example, a Zomato food pic or an IPL player). Generate and display 4 augmented versions.<br><br><em><strong>Hint:</strong> Set rotation_range and horizontal_flip in ImageDataGenerator, then use flow() or flow_from_directory().</em>

```python
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image

# Load Image
img = image.load_img("food.jpg", target_size=(224,224))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)

# Data Augmentation
datagen = ImageDataGenerator(
    rotation_range=30,
    horizontal_flip=True
)

# Generate Images
generator = datagen.flow(img_array, batch_size=1)

plt.figure(figsize=(10,10))

for i in range(4):
    batch = next(generator)

    plt.subplot(2,2,i+1)
    plt.imshow(batch[0].astype("uint8"))
    plt.axis("off")

plt.show()
```

4. Write code to increase and decrease the brightness of a selfie or any image by 40% using ImageDataGenerator's brightness_range parameter. Show the original, brightened, and darkened images side by side.

```python
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image

# Load Image
img = image.load_img("selfi.jpg", target_size=(224,224))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)

# Bright Image
bright_gen = ImageDataGenerator(brightness_range=(1.4,1.4))

# Dark Image
dark_gen = ImageDataGenerator(brightness_range=(0.6,0.6))

bright = next(bright_gen.flow(img_array, batch_size=1))
dark = next(dark_gen.flow(img_array, batch_size=1))

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(img)
plt.title("Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(bright[0].astype("uint8"))
plt.title("Bright +40%")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(dark[0].astype("uint8"))
plt.title("Dark -40%")
plt.axis("off")

plt.show()
```

5. Pick any meme or cricket image and use ImageDataGenerator to apply a shear transformation. Explain in one line how the image changes visually after shearing.

```python
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image

# Load Image
img = image.load_img("virat_kohli.jpg", target_size=(224,224))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)

# Shear Transformation
datagen = ImageDataGenerator(shear_range=30)

generator = datagen.flow(img_array, batch_size=1)

sheared = next(generator)

plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(sheared[0].astype("uint8"))
plt.title("Sheared")
plt.axis("off")

plt.show()
```
