1. Build a simple movie recommendation neural network using Keras Functional API that takes two inputs: user age and favorite genre (as integers), and outputs a predicted movie rating between 1 and 5.



```python
from tensorflow.keras.layers import Input, Dense, Concatenate
from tensorflow.keras.models import Model

# Two Inputs
age_input = Input(shape=(1,), name="User_Age")
genre_input = Input(shape=(1,), name="Favorite_Genre")

# Combine Inputs
merged = Concatenate()([age_input, genre_input])

# Hidden Layers
x = Dense(16, activation="relu")(merged)
x = Dense(8, activation="relu")(x)

# Output Layer
rating_output = Dense(1, activation="linear", name="Movie_Rating")(x)

# Functional API Model
model = Model(
    inputs=[age_input, genre_input],
    outputs=rating_output
)

# Compile Model
model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)

model.summary()
```

2. Create a multi-output neural network using Keras Functional API that predicts both the delivery time (in minutes) and delivery rating (1-5 stars) for a Zomato-style food order, given order amount and distance as inputs.<br><br><em><strong>Hint:</strong> Use two Dense output layers, one for each prediction.</em>

```python
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model

# Input Layer
inputs = Input(shape=(2,), name="Order_Input")

# Hidden Layers
x = Dense(32, activation="relu")(inputs)
x = Dense(16, activation="relu")(x)

# Output 1
delivery_time = Dense(
    1,
    activation="linear",
    name="Delivery_Time"
)(x)

# Output 2
delivery_rating = Dense(
    1,
    activation="linear",
    name="Delivery_Rating"
)(x)

# Functional API Model
model = Model(
    inputs=inputs,
    outputs=[delivery_time, delivery_rating]
)

# Compile Model
model.compile(
    optimizer="adam",
    loss={
        "Delivery_Time": "mse",
        "Delivery_Rating": "mse"
    }
)

model.summary()
```

4. Implement a skip connection in a Keras Functional API model for an IPL cricket score predictor, where the input layer connects both to an intermediate Dense layer and directly to the output layer.<br><br><em><strong>Hint:</strong> Use the 'Add' layer to combine outputs from different layers.</em>

```python
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model

inputs = Input(shape=(2,))

x = Dense(16, activation="relu")(inputs)
x = Dense(8, activation="relu")(x)

outputs = Dense(1, activation="sigmoid")(x)

model = Model(
    inputs=inputs,
    outputs=outputs
)

model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

model.summary()
```
