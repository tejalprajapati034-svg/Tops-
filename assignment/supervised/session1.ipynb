### 4. Imagine you are building a spam filter for WhatsApp messages. Write a short paragraph explaining how traditional programming would approach this problem versus how machine learning would approach it.<br><br><em><strong>Hint:</strong> Focus on rules vs learning from data.</em>

#### Traditional Programming:
- In traditional programming, developers write fixed rules to identify spam messages. For example, if a message contains words like "Win Money", "Click Here", or has many suspicious links, it is marked as spam. Whenever spammers change their tactics, the rules must be updated manually.

#### Machine Learning:
- In machine learning, the spam filter is trained using a large dataset of spam and non-spam WhatsApp messages. Instead of relying on fixed rules, the model learns patterns from the data, such as message content, word frequency, links, and sender behavior. It can detect new types of spam automatically and improves its accuracy as it learns from more data.

#### Difference
- Traditional Programming: Uses predefined rules written by developers.
- Machine Learning: Learns patterns from data and can adapt to new spam messages without manually adding new rules.
