1. Download a small dataset of IPL cricket player stats (e.g., runs, wickets, matches, strike rate) and apply StandardScaler from scikit-learn to scale all numeric features. Save the scaled data to a new CSV file named ipl_scaled.csv.



```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("ipl_players_stats.csv")

# Numeric columns
X = df[["runs","wickets","matches","strike_rate"]]

scaler = StandardScaler()
scaled = scaler.fit_transform(X)

scaled_df = pd.DataFrame(
    scaled,
    columns=X.columns
)

# Save
scaled_df.to_csv("ipl_scaled.csv", index=False)

print(scaled_df.head())
```

2. Using a dataset of Flipkart product ratings (features: price, rating, number of reviews, discount), apply MinMaxScaler so that all features are scaled between 0 and 1. Print the minimum and maximum value for each column before and after scaling.

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("flipkart_products.csv")

X = df[["price","rating","reviews","discount"]]

print("Before Scaling")
print(X.min())
print(X.max())

scaler = MinMaxScaler()

scaled = scaler.fit_transform(X)

scaled_df = pd.DataFrame(
    scaled,
    columns=X.columns
)

print("\nAfter Scaling")
print(scaled_df.min())
print(scaled_df.max())
```

3. Given a Spotify songs dataset (features: danceability, energy, tempo, popularity), use correlation analysis to select the top 2 features most related to 'popularity'. List the selected features and explain your choice in one line each.

```python
import pandas as pd

df = pd.read_csv("spotify.csv")

corr = df.corr()

print(corr)

top = corr["popularity"].drop("popularity").abs().sort_values(ascending=False)

print("\nTop 2 Features")
print(top.head(2))
```

4. Take a Zomato restaurant dataset and use scikit-learn's RFE (Recursive Feature Elimination) with a DecisionTreeRegressor to select the 3 most important features for predicting 'average cost for two'. List the selected features and the RFE ranking.

```python
import pandas as pd

from sklearn.tree import DecisionTreeRegressor
from sklearn.feature_selection import RFE

df = pd.read_csv("zomato.csv")

X = df[["rating","votes","online_order","table_booking"]]

y = df["average_cost_for_two"]

model = DecisionTreeRegressor(random_state=42)

rfe = RFE(model, n_features_to_select=3)

rfe.fit(X,y)

print("Selected Features")

for col, support, rank in zip(X.columns, rfe.support_, rfe.ranking_):
    print(col, support, rank)
```

5. Given a BookMyShow movie ratings dataset, detect outliers in the 'user ratings' column using the IQR method. Remove the outliers and display how many rows were dropped.<br><br><em><strong>Hint:</strong> Calculate Q1, Q3, and IQR; filter out rows outside [Q1 - 1.5*IQR, Q3 + 1.5*IQR].</em>

```python
import pandas as pd

df = pd.read_csv("bookmyshow.csv")

Q1 = df["user_ratings"].quantile(0.25)
Q3 = df["user_ratings"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

clean_df = df[
    (df["user_ratings"] >= lower) &
    (df["user_ratings"] <= upper)
]

print("Original Rows :", len(df))
print("Rows After Removing Outliers :", len(clean_df))
print("Dropped Rows :", len(df) - len(clean_df))
```
