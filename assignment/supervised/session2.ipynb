2. Split the loaded Spotify dataset into training and test sets using sklearn's train_test_split function, with 80% for training and 20% for testing. Print the number of rows in each set.

```python
from sklearn.model_selection import train_test_split
import pandas as pd

# Load dataset
df = pd.read_csv("spotify_top50.csv")

# Features and Label
X = df.drop("popularity", axis=1)
y = df["popularity"]

# Convert categorical columns into numbers
X = pd.get_dummies(X, drop_first=True)

# Split dataset (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Training Samples :", len(X_train))
print("Testing Samples  :", len(X_test))
```

3. Given a simple linear regression model predicting song popularity from danceability, intentionally use only 5% of the data for training and 95% for testing. Observe the model's performance and explain whether this is likely to cause underfitting or overfitting.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load dataset
df = pd.read_csv("spotify_top50.csv")
```

```python
df.shape
```

```python

# Features and Label
X = df[["danceability"]]
y = df["popularity"]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    train_size=0.05,
    test_size=0.95,
    random_state=42
)

model = LinearRegression()
model.fit(X_train, y_train)

# Predictions
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)

# Errors
train_error = mean_squared_error(y_train, train_pred)
test_error = mean_squared_error(y_test, test_pred)

print("Training Error:", train_error)
print("Testing Error :", test_error)
```

4. Draw or find a graph online that visually explains the bias–variance tradeoff, and write a short note describing how this tradeoff would affect predictions in a Zomato restaurant rating predictor app.

#### Explanation (Zomato Example)

- In a Zomato restaurant rating prediction app, a model with high bias is too simple and may predict nearly the same rating for all restaurants, resulting in poor accuracy. A model with high variance is too complex and may memorize the training data, performing well on known restaurants but poorly on new ones. A balanced model provides accurate predictions for both existing and new restaurants.
