2. For the 'player_age' column in your IPL dataset, fill all missing values with the median age using pandas' fillna() method and display the updated column.

```python
import pandas as pd

df = pd.read_csv("ipl_players.csv")

median_age = df["player_age"].median()

df["player_age"] = df["player_age"].fillna(median_age)

print(df["player_age"])
```

3. Suppose the 'team' column in your IPL dataset contains missing values. Replace all missing team names with the constant value 'Unknown' and print the first 10 rows.

```python
import pandas as pd

df = pd.read_csv("ipl_players.csv")

df["team"] = df["team"].fillna("Unknown")

print(df.head(10))
```

4. Take the 'venue' column (categorical) from your IPL dataset and apply one-hot encoding using pandas' get_dummies(). Show the resulting DataFrame with the new columns.

5. Use LabelEncoder from sklearn to encode the 'player_role' column (e.g., Batsman, Bowler, Allrounder) in your IPL dataset, and display the mapping from original roles to encoded values.<br><br><em><strong>Hint:</strong> Use LabelEncoder's classes_ attribute to see the mapping.</em>

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("ipl_players.csv")

encoder = LabelEncoder()

df["player_role_encoded"] = encoder.fit_transform(df["player_role"])

print(df[["player_role", "player_role_encoded"]])

print("\nMapping:")

for role, value in zip(encoder.classes_, range(len(encoder.classes_))):
    print(role, "->", value)
```
