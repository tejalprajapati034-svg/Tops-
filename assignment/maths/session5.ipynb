2. Write a Python script that simulates random sampling: given a list of 1000 fake user IDs (e.g., user1 to user1000), select 50 random users to simulate a survey sample. Print the selected user IDs.<br><br><em><strong>Hint:</strong> Use the random.sample() function from Python's random module.</em>

```python
import random

# Create 1000 fake user IDs
users = [f"user{i}" for i in range(1, 1001)]

# Select 50 random users
sample_users = random.sample(users, 50)

print("Selected Survey Users:\n")

for user in sample_users:
    print(user)
```

4. Given this scenario: A survey about music preferences is only sent to users who recently streamed Bollywood songs on Spotify. Identify and explain the type of bias this introduces in the sample.

#### Scenario: A survey is sent only to users who recently streamed Bollywood songs on Spotify.

- Answer:

- This introduces Selection Bias because only Bollywood listeners are included in the survey. Users who mainly listen to Pop, Rock, Hip-Hop, Classical, or other genres are excluded, so the results do not represent all Spotify users.

5. Choose any two types of sampling (for example: random sampling, stratified sampling, systematic sampling, cluster sampling). For each, describe how you would use it to select a sample of users from BookMyShow to study movie booking habits.

#### A. Random Sampling

- Select users randomly from the complete list of BookMyShow users. Every user has an equal chance of being selected, making the sample fair and unbiased.

#### B. Stratified Sampling

- Divide users into groups based on age (e.g., Under 18, 18–30, 31–50, Above 50). Then randomly select users from each group. This ensures that all age groups are represented in the study of movie booking habits.
