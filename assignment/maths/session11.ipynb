1. Create a Python list to represent a playlist of 5 song durations in seconds (e.g., [210, 180, 245, 200, 175]). Print the total duration of the playlist.

```python
# Playlist durations in seconds
playlist = [210, 180, 245, 200, 175]

# Calculate total duration
total_duration = sum(playlist)

print("Playlist:", playlist)
print("Total Duration:", total_duration, "seconds")
```

2. Represent a 3x3 matrix in Python to store the ratings (1-5) given by 3 friends to 3 different food items on Swiggy. Print the matrix and display the rating each friend gave to the second food item.

```python
# Rows = Friends
# Columns = Food Items

ratings = [
    [5, 4, 3],   # Friend 1
    [4, 5, 2],   # Friend 2
    [3, 4, 5]    # Friend 3
]

print("Ratings Matrix:")

for row in ratings:
    print(row)

print("\nRating for the Second Food Item:")

for i in range(len(ratings)):
    print(f"Friend {i+1}: {ratings[i][1]}")
```

3. Given a matrix representing the number of orders for 4 different dishes across 3 days (rows: days, columns: dishes), write code to calculate and print the total orders for each dish.

```python
# Rows = Days
# Columns = Dishes

orders = [
    [20, 15, 18, 10],   # Day 1
    [22, 18, 20, 12],   # Day 2
    [25, 20, 22, 15]    # Day 3
]

num_dishes = len(orders[0])

print("Total Orders for Each Dish:")

for dish in range(num_dishes):
    total = 0
    for day in orders:
        total += day[dish]
    print(f"Dish {dish+1}: {total}")
```

4. Imagine you have a table of 5 Flipkart products with columns for price and quantity sold. Represent this data as a matrix in Python, then write a function to calculate the total revenue for each product.<br><br><em><strong>Hint:</strong> Multiply price and quantity for each row, then print the results.</em>

```python
# Matrix format:
# [Price, Quantity Sold]

products = [
    [500, 20],
    [1200, 10],
    [800, 15],
    [1500, 8],
    [300, 30]
]

def calculate_revenue(data):
    print("Revenue for Each Product:\n")

    for i in range(len(data)):
        price = data[i][0]
        quantity = data[i][1]
        revenue = price * quantity

        print(f"Product {i+1}: ₹{revenue}")

calculate_revenue(products)
```
