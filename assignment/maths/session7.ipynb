1. Create two Python lists: one representing the number of hours you spend on Instagram each day for a week, and another for the number of posts you make each day. Plot these as a scatter plot using matplotlib and visually describe whether the relationship looks like a positive, negative, or zero correlation.


```python
import matplotlib.pyplot as plt

# Instagram usage (hours per day)
hours = [1, 2, 2.5, 3, 4, 5, 6]

# Number of posts per day
posts = [1, 1, 2, 3, 3, 4, 5]

plt.scatter(hours, posts)

plt.title("Instagram Hours vs Posts")
plt.xlabel("Hours Spent on Instagram")
plt.ylabel("Number of Posts")

plt.grid(True)
plt.show()
```

2. Given the following data for 6 friends: their monthly Zomato order count and their monthly Swiggy order count, calculate the correlation coefficient between the two lists using numpy's corrcoef(). Print the result and explain if the correlation is positive, negative, or zero.

```python
import numpy as np

# Monthly order counts
zomato = [5, 8, 10, 12, 15, 18]
swiggy = [4, 7, 9, 11, 14, 17]

# Correlation coefficient
correlation = np.corrcoef(zomato, swiggy)

print("Correlation Matrix:")
print(correlation)

print("\nCorrelation Coefficient:", correlation[0, 1])

if correlation[0, 1] > 0:
    print("Positive Correlation")
elif correlation[0, 1] < 0:
    print("Negative Correlation")
else:
    print("Zero Correlation")
```
