1. Write a Python function bayes_posterior(prior, likelihood, evidence) that calculates and returns the posterior probability using Bayes theorem formula for given values.

```python
def bayes_posterior(prior, likelihood, evidence):
    """
    Bayes Theorem:
    Posterior = (Likelihood × Prior) / Evidence
    """
    posterior = (likelihood * prior) / evidence
    return posterior


# Example
prior = 0.10
likelihood = 0.80
evidence = 0.26

result = bayes_posterior(prior, likelihood, evidence)

print("Posterior Probability:", round(result, 4))
```

2. Given the following scenario: On Swiggy, 10% of restaurants are 'premium', and 80% of premium restaurants offer a discount, while only 20% of regular restaurants do. If you see a discount, use Bayes theorem in code to find the probability that the restaurant is premium.<br><br><em><strong>Hint:</strong> Define prior, likelihood, and evidence in your code with comments.</em>

```python
# Bayes Theorem Example

# Prior probability:
# 10% restaurants are premium
prior = 0.10

# Likelihood:
# 80% premium restaurants offer discounts
likelihood = 0.80

# Regular restaurants
regular = 0.90

# 20% regular restaurants offer discounts
regular_discount = 0.20

# Evidence:
# Probability of seeing a discount
evidence = (likelihood * prior) + (regular_discount * regular)

# Bayes Theorem
posterior = (likelihood * prior) / evidence

print("Probability restaurant is Premium given Discount =",
      round(posterior, 4))
```

3. Simulate a basic spam detection filter: Given a list of 20 email subjects (some spam, some not), and the probability that 'free' appears in spam vs non-spam emails, write code to calculate the probability an email is spam if it contains 'free', using Bayes theorem.

```python
# Sample email subjects
emails = [
    "Free Gift", "Meeting Today", "Free Offer", "Project Update",
    "Win Cash", "Free Recharge", "Office Notice", "Discount Offer",
    "Free Coupon", "Exam Schedule", "Lottery Winner", "Free Ticket",
    "Team Meeting", "Free Bonus", "Invoice", "Sale Today",
    "Free Trial", "Job Alert", "Free Membership", "Report"
]

# Prior probability of Spam
prior_spam = 0.40

# Probability that the word "free" appears in spam emails
likelihood = 0.90

# Probability that "free" appears in non-spam emails
free_given_not_spam = 0.10

# Evidence
evidence = (likelihood * prior_spam) + \
           (free_given_not_spam * (1 - prior_spam))

# Posterior probability
posterior = (likelihood * prior_spam) / evidence

print("Probability email is Spam if it contains 'free':")
print(round(posterior, 4))
```
