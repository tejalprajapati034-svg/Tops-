
1. Write a Python function calculate_probability(event_count, total_outcomes) that returns the probability of an event, and use it to find the probability of getting 'Heads' when tossing a coin.


```python
def calculate_probability(event_count, total_outcomes):
    return event_count / total_outcomes

# Probability of getting Heads
heads = 1
total_outcomes = 2

probability = calculate_probability(heads, total_outcomes)

print("Probability of getting Heads:", probability)
```

2. Simulate rolling a six-sided dice 100 times using Python's random module, count how many times a '6' appears, and calculate the experimental probability of getting a '6'.<br><br><em><strong>Hint:</strong> Use random.randint(1,6) in a loop and keep a counter for '6'.</em>

```python
import random

rolls = 100
count_six = 0

for i in range(rolls):
    dice = random.randint(1, 6)

    if dice == 6:
        count_six += 1

experimental_probability = count_six / rolls

print("Number of times 6 appeared:", count_six)
print("Experimental Probability of getting 6:", experimental_probability)
```

3. Given an array of genres from your last 20 Spotify songs (like ['pop', 'rock', 'pop', 'hiphop', ...]), write code to calculate the probability of randomly picking a 'pop' song from your recent history.

```python
songs = [
    'pop', 'rock', 'pop', 'hiphop', 'pop',
    'jazz', 'rock', 'pop', 'classical', 'pop',
    'hiphop', 'rock', 'pop', 'pop', 'jazz',
    'rock', 'pop', 'hiphop', 'pop', 'rock'
]

pop_count = songs.count("pop")
total_songs = len(songs)

probability = pop_count / total_songs

print("Total Songs:", total_songs)
print("Pop Songs:", pop_count)
print("Probability of selecting a Pop song:", probability)
```

4. Write a function that, given two arrays — one for users who ordered food on Zomato and one for users who ordered dessert — calculates the probability that a randomly selected user ordered both food and dessert (i.e., intersection over total unique users).

```python
def probability_both(food_users, dessert_users):
    food_set = set(food_users)
    dessert_set = set(dessert_users)

    both = food_set.intersection(dessert_set)
    total_users = food_set.union(dessert_set)

    probability = len(both) / len(total_users)

    return probability


food = ["U1", "U2", "U3", "U4", "U5"]
dessert = ["U2", "U3", "U6", "U7"]

print("Probability of ordering both:",
      probability_both(food, dessert))
```
