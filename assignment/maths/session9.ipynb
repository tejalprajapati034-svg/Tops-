1. Generate 100 random numbers in Python that follow a normal distribution with mean 50 and standard deviation 10 using numpy, then plot the bell curve using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 100 random numbers
data = np.random.normal(loc=50, scale=10, size=100)

# Plot histogram
plt.hist(data, bins=10, density=True, edgecolor='black')

plt.title("Normal Distribution (Mean=50, Std=10)")
plt.xlabel("Values")
plt.ylabel("Density")

plt.grid(True)
plt.show()
```

2. Given an array of 20 daily step counts from your fitness tracker app, calculate the mean, median, and mode, and check if they are approximately equal as expected in a normal distribution.<br><br><em><strong>Hint:</strong> Use numpy.mean(), numpy.median(), and scipy.stats.mode().</em>

```python
import numpy as np
from scipy import stats

steps = [
    6200, 6500, 7000, 7200, 6800,
    7100, 6900, 6700, 7300, 7050,
    6950, 7150, 6850, 7000, 7100,
    6750, 6900, 7200, 7000, 7050
]

mean = np.mean(steps)
median = np.median(steps)
mode = stats.mode(steps, keepdims=True)

print("Mean:", round(mean, 2))
print("Median:", median)
print("Mode:", mode.mode[0])
```

3. Simulate the scores of 200 players in an online game (e.g., PUBG, Free Fire) using a normal distribution with mean 70 and standard deviation 15, then count how many players scored within one standard deviation (between 55 and 85) of the mean.<br><br><em><strong>Hint:</strong> Use the 68-95-99 rule to check your answer.</em>

```python
import numpy as np

# Generate scores
scores = np.random.normal(loc=70, scale=15, size=200)

# Count players within one standard deviation
count = np.sum((scores >= 55) & (scores <= 85))

print("Players within one standard deviation:", count)
print("Percentage:", round((count / 200) * 100, 2), "%")
```

#### Example: Zomato Delivery Times

- Zomato delivery times often follow a normal distribution because most orders are delivered around the average time (for example, 30–40 minutes). Only a few deliveries are much faster or much slower due to factors like traffic, weather, or restaurant delays. This creates a bell-shaped distribution where most observations are concentrated near the mean.
