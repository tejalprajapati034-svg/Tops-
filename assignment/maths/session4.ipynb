1. Given the daily order counts for a Swiggy delivery partner over 10 days: [12, 15, 17, 14, 13, 16, 200, 18, 14, 15], plot a simple graph (hand-drawn or using any tool) and describe whether the data is left skewed, right skewed, or symmetrical. Explain your reasoning based on the shape.

```python
import matplotlib.pyplot as plt

orders = [12, 15, 17, 14, 13, 16, 200, 18, 14, 15]

plt.plot(range(1, 11), orders, marker='o')
plt.title("Swiggy Daily Order Counts")
plt.xlabel("Day")
plt.ylabel("Orders")
plt.grid(True)

plt.show()
```

2. Take the following list of Instagram influencer follower counts (in thousands): [5, 7, 8, 8, 9, 10, 12, 15, 95]. Calculate the mean and median, then state which is higher and what that tells you about the skewness of the data.<br><br><em><strong>Hint:</strong> If the mean is pulled away from the median, the data is skewed in that direction.</em>

```python
import statistics

followers = [5, 7, 8, 8, 9, 10, 12, 15, 95]

mean = statistics.mean(followers)
median = statistics.median(followers)

print("Followers:", followers)
print("Mean:", round(mean, 2))
print("Median:", median)

if mean > median:
    print("The data is Right Skewed.")
elif mean < median:
    print("The data is Left Skewed.")
else:
    print("The data is Symmetrical.")
```

4. Given this dataset of Flipkart product ratings: [3, 3, 4, 4, 4, 5, 5, 5, 5, 5], draw a rough graph and state if the distribution is symmetrical, left skewed, or right skewed. Justify your answer in one line.

```python
import matplotlib.pyplot as plt

ratings = [3, 3, 4, 4, 4, 5, 5, 5, 5, 5]

plt.hist(ratings, bins=[2.5, 3.5, 4.5, 5.5], edgecolor="black")

plt.title("Flipkart Product Ratings")
plt.xlabel("Ratings")
plt.ylabel("Frequency")

plt.show()
```
