```python
import random

class RestaurantSystem:
    def __init__(self):

        self.menu = {
            1: {"name": "Pizza", "price": 250},
            2: {"name": "Burger", "price": 120},
            3: {"name": "Pasta", "price": 180},
            4: {"name": "Sandwich", "price": 90},
            5: {"name": "Coffee", "price": 70}
        }
        self.orders = {}  

    def display_menu(self):
        print("\n------ MENU ------")
        print("Code   Item        Price ()")
        print("-----------------------------")
        for code, item in self.menu.items():
            print(f"{code}   {item['name']:<10}   ₹{item['price']}")
        print("-----------------------------")

    def take_order(self):
        order_items = []
        total_price = 0

        while True:
            try:
                code = int(input("\nEnter item code (0 to finish): "))
                if code == 0:
                    break
                if code not in self.menu:
                    print("Invalid code! Please try again.")
                    continue
                qty = int(input(f"Enter quantity for {self.menu[code]['name']}: "))
                if qty <= 0:
                    print("Quantity must be positive!")
                    continue
                price = self.menu[code]['price'] * qty
                order_items.append((self.menu[code]['name'], qty, price))
                total_price += price
            except ValueError:
                print("Invalid input! Try again.")

        if order_items:
            order_id = random.randint(1000, 9999)
            self.orders[order_id] = {"items": order_items, "total": total_price}
            print(f"\n Order placed successfully! Your Order ID: {order_id}")
            print("Items Ordered:")
            for item in order_items:
                print(f"- {item[0]} x {item[1]} = ₹{item[2]}")
            print(f"Total: ₹{total_price}")
        else:
            print("No items were ordered!")

    def view_order(self):
        try:
            order_id = int(input("Enter Order ID to view: "))
            if order_id in self.orders:
                order = self.orders[order_id]
                print(f"\nOrder ID: {order_id}")
                print("Items Ordered:")
                for item in order['items']:
                    print(f"- {item[0]} x {item[1]} = ₹{item[2]}")
                print(f"Total Amount: ₹{order['total']}")
            else:
                print(" Order not found!")
        except ValueError:
            print("Invalid input!")

    def cancel_order(self):
        try:
            order_id = int(input("Enter Order ID to cancel: "))
            if order_id in self.orders:
                del self.orders[order_id]
                print(f" Order {order_id} has been canceled.")
            else:
                print(" Order not found!")
        except ValueError:
            print("Invalid input!")

    def run(self):
        while True:
            print("\n===== RESTAURANT SYSTEM =====")
            print("1. Display Menu")
            print("2. Take Order")
            print("3. View Order")
            print("4. Cancel Order")
            print("5. Exit")
            choice = input("Enter your choice: ")

            if choice == "1":
                self.display_menu()
            elif choice == "2":
                self.display_menu()
                self.take_order()
            elif choice == "3":
                self.view_order()
            elif choice == "4":
                self.cancel_order()
            elif choice == "5":
                print("Thank you for visiting! Goodbye")
                break
            else:
                print("Invalid choice! Please try again.")

if __name__ == "__main__":
    system = RestaurantSystem()
    system.run()
```

```python
class ATM:
    def __init__(self):
        self.name = None
        self.mobile = None
        self.pin = None
        self.balance = 0
        self.transactions = []

    def activate_card(self):
        print("\n--- Card Activation ---")
        self.name = input("Enter full name: ").strip()

        while True:
            mobile = input("Enter 10-digit mobile number: ").strip()
            if mobile.isdigit() and len(mobile) == 10:
                self.mobile = mobile
                break
            print("Invalid mobile number. Please try again.")

        while True:
            pin1 = input("Create a 4-digit PIN: ").strip()
            pin2 = input("Re-enter PIN for verification: ").strip()
            if pin1 == pin2 and pin1.isdigit() and len(pin1) == 4:
                self.pin = pin1
                break
            print("PINs do not match or invalid. Try again.")

        while True:
            try:
                deposit = float(input("Enter initial deposit (min ₹1000): "))
                if deposit >= 1000:
                    self.balance = deposit
                    self.transactions.append(f"Deposited ₹{deposit}")
                    print(f"Card Activated Successfully! Current Balance: ₹{self.balance}")
                    break
                else:
                    print("Minimum deposit is ₹1000.")
            except ValueError:
                print("Invalid amount. Please enter a number.")

    def verify_pin(self):
        for _ in range(3):  
            entered = input("Enter your 4-digit PIN: ").strip()
            if entered == self.pin:
                return True
            else:
                print("Incorrect PIN.")
        print("Too many failed attempts. Exiting...")
        return False

    def change_pin(self):
        if not self.verify_pin():
            return
        while True:
            new_pin1 = input("Enter new 4-digit PIN: ").strip()
            new_pin2 = input("Re-enter new PIN: ").strip()
            if new_pin1 == new_pin2 and new_pin1.isdigit() and len(new_pin1) == 4:
                self.pin = new_pin1
                print("PIN changed successfully!")
                break
            else:
                print("PINs do not match or invalid.")

    def deposit(self):
        if not self.verify_pin():
            return
        try:
            amount = float(input("Enter amount to deposit: "))
            if amount > 0:
                self.balance += amount
                self.transactions.append(f"Deposited ₹{amount}")
                print(f"Deposited ₹{amount}. Current Balance: ₹{self.balance}")
            else:
                print(" Invalid deposit amount.")
        except ValueError:
            print("Please enter a valid number.")

    def withdraw(self):
        if not self.verify_pin():
            return
        try:
            amount = float(input("Enter amount to withdraw: "))
            if amount > 0:
            
                if amount <= 1000:
                    charge = 0
                elif amount <= 20000:
                    charge = 100
                elif amount <= 100000:
                    charge = 1000
                else:
                    charge = 2000

                total_deduction = amount + charge
                if total_deduction <= self.balance:
                    self.balance -= total_deduction
                    self.transactions.append(f"Withdrew ₹{amount} (Charge ₹{charge})")
                    print(f"Withdrawn ₹{amount} with ₹{charge} charge. Balance: ₹{self.balance}")
                else:
                    print("Insufficient balance.")
            else:
                print("Invalid withdrawal amount.")
        except ValueError:
            print("Please enter a valid number.")

    def check_balance(self):
        if not self.verify_pin():
            return
        print(f"Current Balance: ₹{self.balance}")

    def view_transactions(self):
        if not self.verify_pin():
            return
        print("\n Last 10 Transactions:")
        if not self.transactions:
            print("No transactions yet.")
        else:
            for t in self.transactions[-10:]:
                print("-", t)

    def menu(self):
        while True:
            print("\n--- ATM Main Menu ---")
            print("1. Change PIN")
            print("2. Deposit Money")
            print("3. Withdraw Money")
            print("4. Check Balance")
            print("5. View Last 10 Transactions")
            print("6. Exit")
            
            choice = input("Select an option: ").strip()
            if choice == "1":
                self.change_pin()
            elif choice == "2":
                self.deposit()
            elif choice == "3":
                self.withdraw()
            elif choice == "4":
                self.check_balance()
            elif choice == "5":
                self.view_transactions()
            elif choice == "6":
                print("Thank you for using the ATM. Goodbye!")
                break
            else:
                print("Invalid option, try again.")
                
                
atm = ATM()
atm.activate_card()  
atm.menu()            
```
